# Without Inplace

```python
def rotate(matrix):
    rows=len(matrix)
    cols=len(matrix[0])
    
    list1=[[0 for _ in range(rows)] for _ in range(rows)]
    # count=0
    for i in range(0,len(matrix)):
        for j in range(0,len(matrix[0])):
            # list1.append(matrix[2-j][i])
            list1[i][j]=matrix[2-j][i]
    matrix=list1
    return matrix
```

```python
rotate([[1,2,3],[4,5,6],[7,8,9]])
```
