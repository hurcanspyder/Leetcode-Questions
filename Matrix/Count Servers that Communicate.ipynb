```python
class Solution(object):
    def countServers(self, grid):
        """
        :type grid: List[List[int]]
        :rtype: int
        """
        rows=len(grid)
        cols=len(grid[0])
        hash_map1={}
        hash_map2={}
        for i in range(rows):
            for j in range(cols):
                if grid[i][j]==1:
                    if i not in hash_map1:
                        hash_map1[i]=1
                    else:
                        hash_map1[i]+=1
                    if j not in hash_map2:
                        hash_map2[j]=1
                    else:
                        hash_map2[j]+=1

        count=0
        for i in range(rows):
            for j in range(cols):
                if grid[i][j]==1:
                    if hash_map1[i]>1 or hash_map2[j]>1:
                        count+=1
        return count

                
        
        
```
