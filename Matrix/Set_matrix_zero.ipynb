```python
def set_matrix_zero(matrix):
    
    zeros=[]
    rows=len(matrix)
    cols=len(matrix[0])
    for i in range(0,rows):
        for j in range(0,cols):

            if matrix[i][j]==0:
                zeros.append([i,j])
    for n in range(0,len(zeros)):
        row1=zeros[n][0]
        cols1=zeros[n][1]
        n=0
        m=0
        while n<rows:
            matrix[n][cols1]=0
            n+=1
        while m<cols:
            matrix[row1][m]=0
            m+=1
    return matrix
```
