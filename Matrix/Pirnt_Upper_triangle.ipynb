# Print upper Triangle

```python
def print_traingle(list1):
    rows=len(list1)
    cols=len(list1[0])
    for i in range(0,rows):
        for j in range(0,cols):
            if j>=i:
                print(list1[i][j],end=" ")
            else:
                print("*",end=" ")
        print()
```

```python
print_traingle([[1,2,3],[4,5,6],[7,8,9]])
```

# Print Diagonals

```python
def print_traingle2(list1):
    rows=len(list1)
    cols=len(list1[0])
    for i in range(0,rows):
        for j in range(0,cols):
            if i==j:
                print(list1[i][j],end=" ")
            else:
                print("*",end=" ")
        print()
```

```python
print_traingle2([[1,2,3],[4,5,6],[7,8,9]])
```
