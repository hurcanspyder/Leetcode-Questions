```python
def nextGreaterElement(nums1, nums2):
    """
    :type nums1: List[int]
    :type nums2: List[int]
    :rtype: List[int]
    """
    result=[]
    for i in range(len(nums1)):
        min1=10**10
        flag=True
        b=nums2.index(nums1[i])
        
        for j in range(b,len(nums2)):
            if nums2[j]>nums1[i]:
                result.append(nums2[j])
                flag=False
                break
        if flag:
            result.append(-1)
    return result
```
