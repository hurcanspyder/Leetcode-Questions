# By hash list

```python
def count_characters(string1):
    hash_map=[0]*25
    for i in range(0,len(string1)):
        asci=ord(string1[i])
        index=asci-97
        hash_map[index]+=1
    print(hash_map)


```

```python
count_characters('aaaaaaabbbbbbbbccd')
```

# By dictonary

```python
def count_characters1(string1):
    dict1={}
    for i in range(0,len(string1)):
        if string1[i] not in dict1:
            dict1[string1[i]]=1
        else:
            dict1[string1[i]]+=1
    print(dict1)
```

```python
count_characters1('aabbccddee')
```

# What if two strings are givened and a question is like find out the element present in the string1 which is in string2 means total elements

```python
def count_characters2(string1,string2):
    dict1={}
    for i in range(0,len(string1)):
        if string1[i] not in dict1:
            dict1[string1[i]]=1
        else:
            dict1[string1[i]]+=1
    for i in range(0,len(string2)):
        if string2[i] in dict1:
            print(f"{string2[i]}:{dict1[string2[i]]}")
        else:
            print(f"{string2[i]}!!No it doesn't exist!!")
  
```

```python

count_characters2('aabbccccddee','aceh')
```
