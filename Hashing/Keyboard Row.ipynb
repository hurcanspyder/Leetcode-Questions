```python
first="qwertyuiopQWERTYUIOP"
second="asdfghjklASDFGHJKL"
third="zxcvbnmZXCVBNM"
```

```python


def solve(a):
    hashmap={first:False,second:False,third:False}
    for i in a:
        if i in first:
           
            if hashmap[second]==True or hashmap[third]==True:
                return False
            hashmap[first]=True
        if i in second:
            if hashmap[first]==True or hashmap[third]==True:
                return False
            hashmap[second]=True
        if i in third:
            
            if hashmap[first]==True or hashmap[second]==True:
                return False
            hashmap[third]=True
    return True

def solve2(words):
    list1=[]
    for i in words:
        
        if solve(i)==True :
            
            list1.append(i)
    return list1



        

```

```python
a=["Hello","Alaska","Dad","Peace"]
```

```python
solve2(a)
```
