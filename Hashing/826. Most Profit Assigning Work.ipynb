```python
class Solution(object):
    def maxProfitAssignment(self, difficulty, profit, worker):
        """
        :type difficulty: List[int]
        :type profit: List[int]
        :type worker: List[int]
        :rtype: int
        """
        hash_map={}
        for i in range(0,len(difficulty)):
            if difficulty[i] not in hash_map:
                hash_map[difficulty[i]]=profit[i]
            else:
                hash_map[difficulty[i]]=max(hash_map[difficulty[i]],profit[i])
        difficulty.sort()
        worker.sort()
        hash_map1={}
        maxi=0
        left=0
        for i in worker:
            flag=True
            if i not in hash_map1:
                if left<len(difficulty):
                    while difficulty[left]<=i:
                        hash_map1[i]=max(maxi,hash_map[difficulty[left]])
                        maxi=hash_map1[i]
                        left+=1
                        if left==len(difficulty):
                            break
                        flag=False
                    if flag:
                        hash_map1[i]=maxi
                else:
                    hash_map1[i]=maxi
        count=0
        print(hash_map)
        print(hash_map1)
        for i in worker:
            count+=hash_map1[i]
        return count
```
