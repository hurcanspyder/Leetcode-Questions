**Here we will pre store the number present in our list and then we gonna use them to find out how many numbers are present**

```python
def hash_list(list1,list2):
    hash_list=[0]*11
    for i in list1:
        hash_list[i]+=1
    # print(hash_list)
    for i in list2:
        if i<1 or i>10:
            print("This number is not present in the list")
        else:
            print(hash_list[i])
```

```python
hash_list([1,2,4,2,1,1,4,5,4,5,6,3,4,2,3,5,6,7,4,3,9,9,8,7,7,9,9,10],[2,3,5,6,7,8,9,1,56])
```

# Using dictonary

```python
def hash_list1(list1,list2):
    hash_map={}
    for i in range(0,len(list1)):
        hash_map[list1[i]]=hash_map.get(list1[i],0)+1
    
    for i in list2:
        if i<1 or i>10:
            print("This number is not present in the list")
        else:
            print(hash_map[i])
```

```python
hash_list1([1,2,4,2,1,1,4,5,4,5,6,3,4,2,3,5,6,7,4,3,9,9,8,7,7,9,9,10],[2,3,5,6,7,8,9,1,56])
```
