# Best way it is original Binary search tree concept

Intuition
In a Binary Search Tree (BST), in-order traversal gives values in ascending order.

To solve this problem, we want each node to hold the sum of all nodes with values ≥ current.

This means we should traverse the tree in reverse in-order:
Right → Node → Left, i.e., in descending order.

By keeping a running sum during traversal, we can update each node correctly.

Approach
Initialize a variable self.summ to store the running cumulative sum.

Perform a reverse in-order traversal:
Traverse right subtree.
Update current node with self.summ.
Traverse left subtree.

The right subtree is visited before the node itself, so the largest values are processed first.

Complexity
Time complexity:
O(n)

Space complexity:
O(h)

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def convertBST(self, root):
        """
        :type root: Optional[TreeNode]
        :rtype: Optional[TreeNode]
        """
        self.summ=0
        def func(root):
            if root==None:
                return None
            func(root.right)
            self.summ+=root.val
            root.val=self.summ
            func(root.left)
            return root
        return func(root)
        
        
```
