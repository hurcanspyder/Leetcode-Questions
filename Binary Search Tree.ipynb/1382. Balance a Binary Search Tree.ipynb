```python
class Solution(object):
    def balanceBST(self, root):
        l1 = []

        # inorder to get sorted list
        def inorder(node):
            if not node:
                return
            inorder(node.left)
            l1.append(node)
            inorder(node.right)

        #Build balanced BST using divide and rule method
        def rebuild(l, r):
            if l > r:
                return None
            mid = (l + r) // 2
            root = l1[mid]
            root.left = rebuild(l, mid-1)
            root.right = rebuild(mid+1, r)
            return root

        inorder(root)
        return rebuild(0, len(l1)-1)
```
