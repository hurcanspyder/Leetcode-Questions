```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def isValidBST(self, root):
        """
        :type root: Optional[TreeNode]
        :rtype: bool
        """
        self.list1=[]
        self.ans=True
        def solve(root):
            if not root:
                return 
            solve(root.left)
            if len(self.list1)!=0:
                if root.val<=self.list1[-1]:
                    self.ans=False
                    return 

            
            self.list1.append(root.val)
            solve(root.right)
        solve(root)
        # for i in range(1,len(self.list1)):
        #     if self.list1[i]<=self.list1[i-1]:
        #         return False
        return self.ans


        



        
```
