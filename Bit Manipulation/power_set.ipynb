```python
class Solution(object):
    def subsets(self, nums):
        """
        :type nums: List[int]
        :rtype: List[List[int]]
        """
        n=len(nums)
        result=[]
        total_subset=1<<n
        for j in range(0,total_subset):
            lst=[]
            for i in range(0,n):
                if j & (1<<i)!=0:
                    lst.append(nums[i])
            result.append(lst)
        return result
```
