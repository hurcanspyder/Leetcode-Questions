```python
def change(binary_str):
    decimal = 0
    count = 0
    # reverse the string so that rightmost bit is at index 0
    for digit in binary_str[::-1]:
        decimal += int(digit) * (2 ** count)
        count += 1
    print(decimal)





    
```

```python
change('1101')
```
