⚠️ Works when nums are small (e.g. less than 31 for 32-bit int) because we use bits to represent numbers.

🧠 Summary
Case	Technique	Works For	Notes
1	XOR	One duplicate	Fastest & clean
2	Bitmask	Multiple duplicates, small range	Memory efficient
3	XOR + Set bits	Detect odd/even occurrences	Used in special bit problems

🧩 Case 1: Only one duplicate number (array of 1…n)

Example:
nums = [1, 3, 4, 2, 2]

🔹 Idea

If numbers are from 1 to n (inclusive), and you have one duplicate,
you can use XOR (^) because:

a ^ a = 0

a ^ 0 = a

So, if you XOR all array elements with all numbers 1...n,
the pairs cancel out — only the duplicate remains.

```python
def findDuplicate(nums):
    n = len(nums) - 1
    xor1 = 0
    xor2 = 0

    for num in nums:
        xor1 ^= num

    for i in range(1, n + 1):
        xor2 ^= i

    return xor1 ^ xor2
```

🧩 Case 2: Multiple duplicates (find all)

If you have multiple duplicates, XOR alone won’t work —
but bit masking can help track seen numbers.

💡 Idea

Use bits as a “presence map”:

Bit at position i indicates whether number i has been seen.

💡 Code

```python
def findDuplicates(nums):
    bitmask = 0
    duplicates = []

    for num in nums:
        if bitmask & (1 << num):  # if bit already set → duplicate
            duplicates.append(num)
        else:
            bitmask |= (1 << num)  # mark as seen
    return duplicates
```
