```python
class Solution(object):
    def isSameTree(self, p, q):
        def preorder(root, result):
            if root is None:
                result.append(None)
                return
            result.append(root.val)
            preorder(root.left, result)
            preorder(root.right, result)
        
        a, b = [], []
        preorder(p, a)
        preorder(q, b)
        return a == b
```
