```python
class Solution:
    def distributeCoins(self, root):
        self.ans=0
        def func(root):
            if not root:
                return 0
            left=func(root.left)
            right=func(root.right)
            self.ans+=abs(left)+abs(right)
            return root.val+left+right-1
        func(root)
        return self.ans

        
            
        
```
