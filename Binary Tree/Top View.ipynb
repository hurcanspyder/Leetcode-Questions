```python
def solve(node):
    from collections import deque
    queue=deque()
    my_dict={}
    ans=[]
    
    queue.append((node,0))
    
    while queue:
        e,line=queue.popleft()
        if line not in my_dict:
            my_dict[line]=e.val
        if e.left is not None:
            
            queue.append(e.left,line-1)
        if e.right is not None:
            
            queue.append(e.right,line+1)
    for value in sorted(my_dict.items()):
        ans.append(value[1])
    return ans
            
            
```
