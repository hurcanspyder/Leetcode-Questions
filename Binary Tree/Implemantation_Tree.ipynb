```python
class Tree():
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

    def insert(self, root, val):
        # Base case
        if root is None:
            return Tree(val)

        if val < root.val:
            root.left = self.insert(root.left, val)

        elif val > root.val:
            root.right = self.insert(root.right, val)

        return root

root=Tree(20)
root.left=Tree(10)
root.right=Tree(30)
root.insert(root,15)



    
```
