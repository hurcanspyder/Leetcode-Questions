```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def goodNodes(self, root):
        """
        :type root: TreeNode
        :rtype: int
        """
        self.count=0
        def func(root,maxi):
            if not root:
                return 
            if root.val>=maxi:
                self.count+=1
            maxi=max(root.val,maxi)
            func(root.left,maxi)
            func(root.right,maxi)
            return 
        func(root,-10**7)
        return self.count  
```
