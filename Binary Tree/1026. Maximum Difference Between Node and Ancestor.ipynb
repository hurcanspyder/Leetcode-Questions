```python
def func(root):
    if not root:
        return 0,10**8,-10**7
    left,left_min,left_max=func(root.left)
    right,right_min,right_max=func(root.right)
    curr_min=min(left_min,right_min,root.val)
    curr_max=min(right_max,left_max,root.val)

    return max(left,right,abs(root.val-curr_min),abs(root.val-curr_max)),curr_min,curr_max


```
