```python
class Node:
    def __init__(self,val):
        self.val=val
        self.left=None
        self.right=None

drinks=Node("drinks")
hot=Node("hot")
cold=Node("cold")
tea=Node("tea")
coffee=Node("coffee")
cola=Node("cola")
fanta=Node("fanta")

hot.left=tea
hot.right=coffee

cold.left=cola
cold.right=fanta

drinks.left=hot
drinks.right=cold
```

```python
print(drinks.left.left.val)
```

```python
def preorder(node):
    if node==None:
        return 
    print(node.val)
    preorder(node.left)
    preorder(node.right)
```

```python
preorder(drinks)
```
