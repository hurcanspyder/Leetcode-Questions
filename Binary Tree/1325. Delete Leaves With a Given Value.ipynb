```python
class Solution(object):
    def removeLeafNodes(self, root, target):
        def func(root):
            if not root:
                return True   # Treat null as removable

            left = func(root.left)
            right = func(root.right)

            if left:
                root.left = None
            if right:
                root.right = None

            # Now check CURRENT node after children are processed
            if root.val == target and not root.left and not root.right:
                return True

            return False

        if func(root):
            return None
        return root

        
```
