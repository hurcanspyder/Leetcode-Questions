# BFS APproach

```python
from collections import deque

class Solution(object):
    def isSymmetric(self, root):
        if not root:
            return True

        queue = deque([root])

        while queue:
            level = []
            for _ in range(len(queue)):
                node = queue.popleft()
                if node != "a":   # it's a TreeNode
                    level.append(node.val)
                    queue.append(node.left if node.left else "a")
                    queue.append(node.right if node.right else "a")
                else:             # it's a placeholder
                    level.append("a")

            if level != level[::-1]:
                return False

        return True
        



```

# DFS approach

```python
class Solution(object):
    def isSymmetric(self, root):
        """
        :type root: Optional[TreeNode]
        :rtype: bool
        """
        if not root:
            return True

        def isMirror(t1, t2):
            if not t1 and not t2:   # both empty → symmetric
                return True
            if not t1 or not t2:    # one empty → not symmetric
                return False
            return (t1.val == t2.val and 
                    isMirror(t1.left, t2.right) and 
                    isMirror(t1.right, t2.left))

        return isMirror(root.left, root.right)
```
