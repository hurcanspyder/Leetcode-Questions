```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def flipEquiv(self, root1, root2):
        """
        :type root1: Optional[TreeNode]
        :type root2: Optional[TreeNode]
        :rtype: bool
        """
        if not root1 or not root2:
            return root1 is root2
        isLeaf1 = not root1.left and not root1.right
        isLeaf2 = not root2.left and not root2.right
        sameVal = root1.val == root2.val
        if isLeaf1:
            return sameVal and isLeaf2
        if isLeaf2:
            return sameVal and isLeaf1
        l = self.flipEquiv(root1.left, root2.right) or self.flipEquiv(root1.left, root2.left)
        r = self.flipEquiv(root1.right, root2.left) or self.flipEquiv(root1.right, root2.right)
        return sameVal and l and r
```
