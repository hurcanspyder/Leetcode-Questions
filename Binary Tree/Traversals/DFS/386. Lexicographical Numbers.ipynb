```python
class Solution(object):
    def lexicalOrder(self, n):
        res = []
        curr = 1
        
        for _ in range(n):
            res.append(curr)
            
            # try to go deeper: multiply by 10
            if curr * 10 <= n:
                curr *= 10
            else:
                # otherwise increase
                while curr % 10 == 9 or curr + 1 > n:
                    curr //= 10   # go up the tree
                curr += 1
        
        return res
```
