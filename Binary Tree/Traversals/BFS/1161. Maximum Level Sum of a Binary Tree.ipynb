```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def maxLevelSum(self, root):
        """
        :type root: Optional[TreeNode]
        :rtype: int
        """
        from collections import deque
        list1=[]
        queue=deque()
        queue.append(root)

        while len(queue)!=0:
            size=len(queue)
            list2=[]
            for i in range(size):
                e=queue.popleft()
                list2.append(e.val)
                if e.left:
                    queue.append(e.left)
                if e.right:
                    queue.append(e.right)
            list1.append(sum(list2))
        maxi=max(list1)
        return list1.index(maxi)+1
        
            
            
```
