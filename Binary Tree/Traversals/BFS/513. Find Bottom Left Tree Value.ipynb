```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def findBottomLeftValue(self, root):
        """
        :type root: Optional[TreeNode]
        :rtype: int
        """
        if not root:
            return 0
        from collections import deque
        queue=deque()
        queue.append(root)
        height=0
        result=[]
        while queue:
            size=len(queue)
            list1=[]
            for i in range(size):
                e=queue.popleft()
                list1.append(e.val)
                if e.left:
                    queue.append(e.left)
                if e.right:
                    queue.append(e.right)
            result.append(list1)
        return result[-1][0]


                
            
        
```
