# Recursive approach

```python
def solve(node):
    if node==None:
        return 0
    LH=solve(node.left)
    RH=solve(node.right)
    return 1+min(LH,RH)
```

# Level Order approach

```python
def level_order(root):
    from collections import deque
    queue=deque([])
    height=0
    queue.append(root)
    while len(queue)!=0:
        level_size=len(queue)
        height+=1
        for _ in range(level_size):
            e=queue.pop()
            if e.left is not None:
                queue.append(e.left)
            if e.right is not None:
                queue.append(e.right)
    return height
```
