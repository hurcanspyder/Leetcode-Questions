```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def pruneTree(self, root):
        """
        :type root: Optional[TreeNode]
        :rtype: Optional[TreeNode]
        """
        def func(root):
            if not root:
                return False
            
            left=func(root.left)
            right=func(root.right)
            if not left:
                root.left=None
            if not right:
                root.right=None
            # if left or right:
            #     return True
            # if not left and not right:
            #     root=None
            #     return False
            return root.val==1 or left or right
        return root if func(root) else None
        

            
```
