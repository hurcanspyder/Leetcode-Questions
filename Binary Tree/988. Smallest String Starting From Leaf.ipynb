```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def smallestFromLeaf(self, root):
        """
        :type root: Optional[TreeNode]
        :rtype: str
        """
        self.ans=None
        def func(root,subset,result):
            
            if not root:
                s="".join(subset[::-1])
                if self.ans is None or s<self.ans:
                    self.ans=s
                return self.ans
            subset.append(chr(root.val+97))
            func(root.left,subset,result)
            
            func(root.right,subset,result)
            subset.pop()
            return self.ans
        func(root,[],[])
        return self.ans

        
```
