# 🧾 Bit Manipulation Cheat Sheet (Python Examples)

## 🔹 1. Basic Operations

```
a, b = 5, 3  # binary: 0101 , 0011

print(a & b)   # AND → 1  (0001)
print(a | b)   # OR  → 7  (0111)
print(a ^ b)   # XOR → 6  (0110)
print(~a)      # NOT → -6 (2's complement)
print(a << 1)  # Left shift → 10 (1010) = a * 2
print(a >> 1)  # Right shift → 2 (0010) = a // 2
```

## 🔹 2. Checking Even/Odd

```
n = 7
if n & 1:   # last bit 1 → odd
    print("Odd")
else:
    print("Even")
```

## 🔹 3. Swap Two Numbers (XOR Trick)

```
a, b = 10, 20
a = a ^ b
b = a ^ b
a = a ^ b
print(a, b)  # 20, 10
```

## 🔹 4. Check if a Number is Power of 2

```
def isPowerOfTwo(n):
    return n > 0 and (n & (n - 1)) == 0

print(isPowerOfTwo(16))  # True
print(isPowerOfTwo(18))  # False
```

## 🔹 5. Count Set Bits (1s in Binary)

```
def countBits(n):
    count = 0
    while n:
        n &= (n - 1)
        count += 1
    return count

print(countBits(29))  # 29 = 11101 → 4
```

## 🔹 6. Turn ON / OFF / TOGGLE a Bit

```
n = 10   # 1010
i = 1    # 2nd bit (0-indexed from right)

# Set bit (make it 1)
n |= (1 << i)  
print(bin(n))  # 0b1010 → stays 1010 (already 1)

# Clear bit (make it 0)
n &= ~(1 << i)
print(bin(n))  # 0b1000

# Toggle bit (flip 0 ↔ 1)
n ^= (1 << i)
print(bin(n))  # 0b1010
```

## 🔹 7. Get ith Bit

```
def getBit(n, i):
    return (n >> i) & 1

print(getBit(10, 1))  # 10 = 1010 → 1st bit = 1
```

## 🔹 8. Subset Generation (Bitmasking)

```
arr = [1, 2, 3]
n = len(arr)

for mask in range(1 << n):   # 0 to 2^n - 1
    subset = []
    for i in range(n):
        if mask & (1 << i):
            subset.append(arr[i])
    print(subset)
```

## 🔹 9. Find Single Number in Array (all others twice)

```
arr = [2, 3, 2, 4, 4]
xor = 0
for num in arr:
    xor ^= num
print(xor)  # 3
```

## 🔹 10. Find Two Unique Numbers (all others twice)

```
arr = [2, 4, 7, 9, 2, 4]
xor = 0
for num in arr:
    xor ^= num

# Get rightmost set bit
set_bit = xor & -xor

x, y = 0, 0
for num in arr:
    if num & set_bit:
        x ^= num
    else:
        y ^= num

print(x, y)  # 7, 9
```

## 🔹 11. Fast Multiply / Divide by 2

```
n = 10
print(n << 1)  # 20 (multiply by 2)
print(n >> 1)  # 5  (divide by 2)
```
