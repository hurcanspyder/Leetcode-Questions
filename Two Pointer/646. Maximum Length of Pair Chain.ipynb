```python
class Solution(object):
    def findLongestChain(self, pairs):
        """
        :type pairs: List[List[int]]
        :rtype: int
        """
        pairs.sort(key=lambda x:x[1])
       
        count=0
        last=-10**8
        for x,y in pairs:
            if x>last:
                count+=1
                last=y
        return count
```
