```python
class Solution(object):
    def numSubarrayBoundedMax(self, nums, left, right):
        
        def count_subarrays(bound):
            count = 0
            length = 0
            for x in nums:
                if x <= bound:
                    length += 1
                    count += length
                else:
                    length = 0
            return count
        
        return count_subarrays(right) - count_subarrays(left - 1)




           

        
```
