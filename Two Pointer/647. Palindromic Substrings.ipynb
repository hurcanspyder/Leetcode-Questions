```python
class Solution(object):
    def countSubstrings(self, s):
        """
        :type s: str
        :rtype: int
        """
        count=0
        for i in range(0,len(s)):
            for j in range(i+1,len(s)):
                if i==0:
                    if s[i:j+1]==s[j::-1]:
                        count+=1
                else:
                    if s[i:j+1]==s[j:i-1:-1]:
                        count+=1
        return count+len(s)
```
