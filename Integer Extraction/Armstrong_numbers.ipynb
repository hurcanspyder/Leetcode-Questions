# Armstrong Number
Armstrong number is defined as for eg.number is abc so the the sum of a^3+b^3+c^3 should be equal to abc

means

abc= a^3+b^3+c^3 

```python
def check_armstrong_number(nums):
    result=0
    n=nums
    nods=len(str(n))
    while(nums>0):
        result=(nums%10)**nods+result
        nums=nums//10
    
    if result==n:
        print("YES")
    else:
        print("NO")
```

```python
check_armstrong_number(153)
```

**Time Complexity for this code is o(log10(N)) because we are dividing it by 10 after every loop**
