```python
def count_digitts(nums):
    count=0
    while(nums>0):
        count+=1
        nums=nums//10
    print(count)



    
```

```python
count_digitts(5873)
```

**Here time complexity is o(log10n)**

# From log methods

```python
def count_digitts1(nums):
    from math import log10
    print(int(log10(nums)+1))
```

```python
count_digitts1(5873)
```

**Here time complexity is constant means O(1)**
