# Method 1

```python
def frequency_map(nums):
    freq_map={}
    for i in range(0,len(nums)):
        if nums[i] in freq_map:
            freq_map[nums[i]]+=1
        else:
            freq_map[nums[i]]=1
    print(freq_map)

frequency_map([5,6,7,4,3,2,2,3,3,3,3,3,2])
```

**Here the time complexity of our code is O(n) because to search the element in the dictonary take constant time complexit**


# Method 2

```python
def frequency_map(nums):
    hash_map={}
    for i in range(0,len(nums)):
        hash_map[nums[i]]=hash_map.get(nums[i],0)+1

frequency_map([5,6,7,4,3,2,2,3,3,3,3,3,2])
```
