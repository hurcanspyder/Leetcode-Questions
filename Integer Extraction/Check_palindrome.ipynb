```python
def check_palindrome(nums):
    result=0
    n=nums
    while(nums>0):
        result=result*10+nums%10
        nums=nums//10
    if(n==result):
        print("YES")
    else:
        print("NO")


nums=int(input())
check_palindrome(nums)
```

**Time Complexity is O(log10(n))**

**Constant space complexity O(N)**
