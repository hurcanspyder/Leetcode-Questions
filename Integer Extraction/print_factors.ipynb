```python
def print_factors(nums):
    result=[]
    for i in range(1,nums+1):
        if nums%i==0:
            result.append(i)
            
    print(result)
```

```python
print_factors(20)
```

# The above code is bruteforce approach
because the time comlexity if O(n) and space compllexity is O(k) k would be the total number of factors 


# Optimal Approach

```python
def print_factors1(nums):
    result=[]
    for i in range(1,nums//2):
        if nums%i==0:
            result.append(i)
    result.append(nums)
    print(result)
```

```python
print_factors1(20)
```

**Here we decresed the time complexity by O(n/2) because after n/2 there is not even a single number except n who can divide n**

# More optimal approach


```python
def print_factors2(nums):
    
    result=[]
    for i in range(1,int(nums**(1/2))+1):
        if nums%i==0:
            result.append(i)
            if nums//i!=i:
                result.append(nums//i)
    
    print(result)
```

```python
print_factors2(16)
```

By above code we decreased rge time complexity by O(n**(1/2)) which is good then n/2
