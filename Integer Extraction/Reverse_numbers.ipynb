```python
def reverse_number(nums):
    
    result=0
    if nums>0:
        while(nums>0):
            result=result*10+nums%10
            nums=nums//10
        if -2**31<=result<=2**31-1:
            return result
        else:
            return 0
    else:
        nums=nums*-1
        while(nums>0):
            result=result*10+nums%10
            nums=nums//10
        
        if -2**31<=-result<=2**31-1:
            return -result

        else:
            return 0


    
    


nums=int(input())
reverse_number(nums)
```
