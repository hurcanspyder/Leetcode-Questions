✅ What is Sieve of Eratosthenes?

A fast algorithm to find all prime numbers up to n-1 in O(n log log n) time.

Key idea:

For every prime p, mark all multiples of p as composite starting from p*p.

Because smaller multiples like 2*p, 3*p were already marked by smaller primes.

# Mine one

```python
class Solution(object):
    def countPrimes(self, n):
        """
        :type n: int
        :rtype: int
        """
        if n<=2:
            return 0
        dp=[True for _ in range(n)]

        dp[0]=False
        dp[1]=False
        count=0
        for i in range(2,n):
            if dp[i]:
                for j in range(i*i,n,i):
                    dp[j]=False
                count+=1
        return count

        
```
