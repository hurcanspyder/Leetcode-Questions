```python
class Solution(object):
    def detectCapitalUse(self, word):
        """
        :type word: str
        :rtype: bool
        """
        up=0
        low=0


        for i in word:
            if ord(i)>=97 and ord(i)<=122:
                low+=1
            if ord(i)>=65 and ord(i)<=90:
                up+=1
        if up ==len(word):
            return True
        if up==1 and (ord(word[0])>=65 and ord(word[0])<=90):
            return True
        if low==len(word):
            return True
        return False


```
