```python
def remove_duplicates1(list1):
    
    list2=[]
    count=0
    for i in range(0,len(list1)):
        if list1[i] not in list2:
           
            list2.append(list1[i])
            count+=1
    list2=list2+[0]*(len(list1)-count)
    list1=list2
    print(list1)
   
```

```python
remove_duplicates1([1,2,2,2,2,2,3,3,3,4,5,5,5,6,6])
```

# By two Pointers approach

```python
def dupliacte(list1):
    i=0
    n=len(list1)
    if n==1:
        return 1
    for j in range (1,len(list1)):
        if list1[i]!=list1[j]:
            i+=1
            temp=list1[j]
            list1[i]=list1[j]
            list1[j]=temp
    print(list1)
```

```python
dupliacte([1,2,2,2,2,2,3,3,3,4,5,5,5,6,6])
```

**this is best approach working on O(n) time complexity and O(1) space complexity**
