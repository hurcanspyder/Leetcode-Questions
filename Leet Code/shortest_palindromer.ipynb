```python
def shortest_palindrome(s):
    a1=s[::-1]
    i=0
    j=0
    result=""
    while(i<len(a1) and j<len(s)):
        if result[::-1]==result and len(result)>1:
            print("Hello")
            return result
        elif a1[i]!=s[j]:
            result=result+a1[i]
            i+=1
        
        else:
            result=result+a1[i]
            i+=1
            j+=1
        print(result)
    while i<len(a1):
        if result[::-1]==result:
            return result
        result=result+a1[i]
        i+=1
    while j<len(s):
        if result[::-1]==result:
            return result
        result=result+s[j]
        j+=1
    return result
```
