```python
class Solution(object):
    def doesAliceWin(self, s):
        """
        :type s: str
        :rtype: bool
        """
        count=0
        vowels={'a','e','i','o','u'}
        for i in range(len(s)):
            if s[i] in vowels:
                count+=1
        if count==0:
            return False
        elif count%2==0:
            return True 
        else:
            return True
```
