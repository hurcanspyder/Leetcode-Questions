```python
class Solution(object):
    def isPossibleDivide(self, nums, k):
        if len(nums) % k != 0:
            return False

        nums.sort()

        hash_map = {}
        for i in nums:
            hash_map[i] = hash_map.get(i, 0) + 1

        sorted_keys = sorted(hash_map.keys())

        for key in sorted_keys:
            freq = hash_map[key]

            if freq > 0:
                for i in range(k):
                    if hash_map.get(key + i, 0) < freq:
                        return False
                    hash_map[key + i] -= freq

        return True
```
