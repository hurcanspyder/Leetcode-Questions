# By Slicing

```python
def right_rotate(list1):
    list1[:]=[list1[-1]]+list1[0:len(list1)-1]
    print(list1)
```

```python
right_rotate([1,2,3,4,5,6,7,8,9])
```

# Without SLicing

```python
def rotate(list1):
    n=len(list1)
    temp=list1[n-1]
    for i in range(n-2,-1,-1):
        list1[i+1]=list1[i]
    list1[0]=temp
    print(list1)
```

```python
rotate([1,2,3,4,5,6,7,8,9])
```
