# BruteForce approach

```python
def second_largest(list1):
    largest=list1[0]
    list2=[]
    for i in list1:
        if i>largest:
            largest=i
            list2.append(largest)
    print(list2[-2])
    
```

**Time complexity of this code is O(nlogn) and space complexity is O(1)**

```python
second_largest([1,2,3,4,5,-101,3,3,4,44,3,2,2,6,8,99,100])
```

# 2nd approach

```python
def second_largest1(list1):
    largest=-10**300
    second_largest=-10**300
    for i in list1:
        if i>largest:
            largest=i
    for i in list1:
        if i>second_largest and i<largest:
            second_largest=i
    print(second_largest)

   
```

```python
second_largest1([1,2,3,4,5,-101,3,3,4,44,3,2,2,6,8,99,100])
```

**Here the time complexity is O(2n) and space complexity is o(1)**

# optimal Solution

```python
def second_largest2(list1):
    largest=-10**300
    second_largest=-10**300
    for i in list1:
        if i>largest:
            second_largest=largest
            largest=i
        elif i>second_largest and i<largest:
            second_largest=i
    print(second_largest)

```

```python
second_largest2([1,2,3,4,5,-101,3,3,4,44,3,2,2,6,8,99,100])
```

**Here the time complexity is o(n) and space complexity is o(1)**
