```python
def solve(n):
    def change(n,a):
        if n>1:
            change(n//2,a)
        a.append(n%2)
        return a
    b=change(n,[])
    print(b)
    if len(b)==1:
        if b[0]==1:
            return True
        else: False
    count=0
    flag=False
    for i in range(len(b)):
        if b[i]==1:
            if count==0:
                flag=True
                count+=1
            else:
                flag=False
                return flag
    return flag

        
```
