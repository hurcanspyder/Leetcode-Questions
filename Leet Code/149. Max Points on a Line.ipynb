```python
class Solution(object):
    def maxPoints(self, points):

        n = len(points)
        if n <= 2:
            return n
        
        max_points = 2
        
        for i in range(n):
            x1, y1 = points[i]
            
            for j in range(i+1, n):
                x2, y2 = points[j]
                
                count = 2
                
                for k in range(n):
                    if k == i or k == j:
                        continue
                        
                    x3, y3 = points[k]
                    
                    # check collinear using cross multiplication
                    if (y3 - y1) * (x2 - x1) == (y2 - y1) * (x3 - x1):
                        count += 1
                        
                max_points = max(max_points, count)
                
        return max_points



        
```
