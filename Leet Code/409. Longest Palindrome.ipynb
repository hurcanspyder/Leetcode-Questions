```python
class Solution(object):
    def longestPalindrome(self, s):
        hash_map = {}

        for ch in s:
            hash_map[ch] = hash_map.get(ch, 0) + 1

        length = 0
        odd_found = False

        for freq in hash_map.values():
            if freq % 2 == 0:
                length += freq
            else:
                length += freq - 1
                odd_found = True

        return length + 1 if odd_found else length
```
