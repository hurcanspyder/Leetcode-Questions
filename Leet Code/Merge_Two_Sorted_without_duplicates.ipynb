```python
def merge(left,right):
    i=0
    j=0
    result=[]
    while i<len(left) and j<len(right):
        if left[i]<right[j] and left[i] not in result:
            result.append(left[i])
            i+=1
        elif left[i]>right[j] and right[j] not in result:
            result.append(right[j])
            j+=1
        else:  # equal case
            if left[i] not in result:
                result.append(left[i])
            i += 1
            j += 1
    while i<len(left):
        if left[i] not in result:
            result.append(left[i])
            i+=1     
    while j<len(right):
        if right[j] not in result:
            result.append(right[j])
            j+=1
    print(result)         
```

```python
merge([1,1,1,1],[2,2,2,2])
```
