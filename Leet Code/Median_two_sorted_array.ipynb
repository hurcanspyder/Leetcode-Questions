```python
def findMedianSortedArrays( nums1, nums2):
    nums3=[]
    i,j=0,0
    while i<len(nums1) and j<len(nums2):
        if nums1[i]<nums2[j]:
            nums3.append(nums1[i])
            i+=1
        else:
            nums3.append(nums2[j])
            j+=1
    while i<len(nums1):
        nums3.append(nums1[i])
        i+=1

    while j<len(nums2):
        nums3.append(nums2[j])
        j+=1
    if len(nums3)%2==0:
        return (nums3[len(nums3)//2-1]+nums3[len(nums3)//2])/2
    else:
        return nums3[len(nums3)//2]



   
```
