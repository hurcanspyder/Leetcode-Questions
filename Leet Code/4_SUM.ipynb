```python
def sum(list1,target):
    my_set=set()
   
    for i in range(0,len(list1)):
        for j in range(i+1,len(list1)):
            a=set()
            for k in range(j+1,len(list1)):
                diff=target-list1[i]-list1[j]-list1[k]
                if diff in a:
                    list2=[list1[i],list1[j],list1[k],diff]
                    list2.sort()
                    list2=tuple(list2)
                    my_set.add(list2)
                else:
                    a.add(list1[k])
    return list(my_set)

```
