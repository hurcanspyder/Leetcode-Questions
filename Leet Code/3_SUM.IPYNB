```python
def sum(list1):
    my_set=set()
    for i in range(0,len(list1)):
        for j in range(i+1,len(list1)):
            for k in range(j+1,len(list1)):
                if list1[i]+list1[j]+list1[k]==0:
                    list3=[list1[i],list1[j],list1[k]]
                    list3.sort()
                    list3=tuple(list3)
                    my_set.add(list3)
    return my_set
```

```python
sum([-1,0,1,2,-1,-4])
```
