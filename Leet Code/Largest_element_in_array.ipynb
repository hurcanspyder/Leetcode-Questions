```python
def largest_element(list1):
    largest=0
    for i in list1:
        if i>largest:
            largest=i
    print(largest)

```

```python
largest_element([1,2,3,4,5,-101,3,3,4,44,3,2,2,6,8,99,100])
```

*Now this code have one limitation which is what if every element present in our array is less than 0 or negatiive then the value of largest will always be 0 where as 0 is not their in the list*

To do this we will assume largest number to be the first element of the list and then traverse the whole list

```python
def largest_element(list1):
    largest=list1[0]
    for i in list1:
        if i>largest:
            largest=i
    print(largest)
```

```python
largest_element([-1,-2,-3,-4,-5,-6,-0.1,-0.001])
```

**Time complexity of this code is O(N)**
