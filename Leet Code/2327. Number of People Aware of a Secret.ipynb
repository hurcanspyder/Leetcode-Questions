```python
def peopleAwareOfSecret(n, delay, forget):
    hash_table = {1: 1}   # person 1 knows secret at day 1
    count = 1
    new_one = 1

    while count <= n:
        for key, value in list(hash_table.items()):
            # update how many days the person has known the secret
            hash_table[key] = value + 1

            # person can start sharing the secret after delay days
            if delay <= hash_table[key] - 1 < forget:
                new_one += 1
                hash_table[new_one] = 1

            if count == n:
                return len(hash_table)  
            # forget secret after "forget" days
            if hash_table[key] >= forget:
                hash_table.pop(key)

        # if count == n:
        #     return len(hash_table)

        count += 1

```

```python
peopleAwareOfSecret(5,2,3)
```
