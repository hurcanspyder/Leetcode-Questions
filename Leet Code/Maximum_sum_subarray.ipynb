```python
def maximum_subarray(list1):
    total=-10**300
    max_sum=-10**300
    i=0
    j=0
    if len(list1)==1:
        return list1[i]
    while i<len(list1):
        j=i
        total=0
        while j<len(list1):
            
            total=total+list1[j]
            if total>max_sum:
                max_sum=total
            j+=1
        i+=1
    return max_sum

```

```python
maximum_subarray([-2,1])
```

# Kadane's Algorithm

```python
def maximum_subarray1(list1):
    max_sum=-10**300
    total=0
    for i in range(0,len(list1)):
        total=total+list1[i]
        if total>max_sum:
            max_sum=total
        if total<0:
            total=0
    print(max_sum)
```

```python
maximum_subarray1([-2,1,-3,4,-1,2,1,-5,4])
```
