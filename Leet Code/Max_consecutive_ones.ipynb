```python
def findMaxConsecutiveOnes(nums):
    hash_map={}
    list1=nums
    a=0
    b=0
    if len(nums)==1:
        if nums[0]==1:
            return 1
        else:
            return 0
    for i in range(0,len(list1)):
        if list1[i]==1:
            a+=1
            hash_map[b]=a
        else:
            a=0
            b+=1
    if hash_map:
        return max(hash_map.values())
    else:
        return 0
```

```python
print(findMaxConsecutiveOnes([0,0]))
```
