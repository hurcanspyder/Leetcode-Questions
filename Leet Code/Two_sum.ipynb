```python
def two_sum(list1,target):
    for i in range(0,len(list1)):
        for j in range(i+1,len(list1)):
            if list1[i]+list1[j]==target:
                print(i,j)
                
```

```python
two_sum([1,2,4,5,7,4,3,3],6)
```

# Optimal Approach

```python
def two_sum1(list1,target):
    dict={}
    index=0
    for i in range(0,len(list1)):
        difference=target-list1[i]
        if difference in dict:
            return dict[difference],i
        else:
            dict[list1[i]]=index
            index+=1



        
```

```python
two_sum1([3,3],6)
```
