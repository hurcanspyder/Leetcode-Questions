```python
def permute(nums):
    result=[]
    inside=[]
    for i in range(0,len(nums)):
        inside=[]
        inside.append(nums[i])
        for j in range(0,len(nums)):
            if nums[j]==nums[i]:
                continue
            else:
                inside.append(nums[j])
        result.append(inside)
    print(result)
        


```

```python
permute([1,2,3])
```
