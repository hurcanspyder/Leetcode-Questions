```python
def minOperations(queries):
    count = 0
    for l, r in queries:
        nums = list(range(l, r+1))
        left, right = 0, len(nums)//2

        while nums[-1] != 0 and right < len(nums):

            # --- Skip zeros in sync ---
            if nums[left] == 0 and nums[right] == 0:
                left += 1
                right += 1
                continue

            if nums[left] == 0:
                left += 1
                continue

            if nums[right] == 0:
                right += 1
                continue

            # --- When both pointers active ---
            if right == left and right != len(nums)-1:
                # process single element (middle case)
                nums[left] //= 4
                left += 1
                right += 1
                count += 1

            elif right == left and right == len(nums)-1:
                # last element
                nums[left] //= 4
                left += 1
                right += 1
                count += 1

            else:
                # process pair
                nums[left] //= 4
                nums[right] //= 4

                if right == len(nums)-1 and nums[right] != 0:
                    left = 0
                    right = len(nums)//2
                else:
                    left += 1
                    right += 1
                count += 1

    return count

```
