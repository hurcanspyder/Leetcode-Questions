```python
class Solution(object):
    def validPalindrome(self, s):
        """
        :type s: str
        :rtype: bool
        """
        n=len(s)
        if s==s[::-1]:
            return True
        else:
            for i in range(0,len(s)):
                if i==0:
                    if s[1::]==s[:0:-1]:
                        return True
                else:
                    if s[0:i]+s[i+1:n]==s[:i:-1]+s[i-1::-1]:
                        return True
        return False
    
```
