```python
def maxFreqSum( s):
    """
    :type s: str
    :rtype: int
    """
    vowels={'a','e','i','o','u'}
    my_vowel={}
    my_consonants={}
    for i in range(len(s)):
        if s[i] in vowels:
            if s[i] not in my_vowel:
                my_vowel[s[i]]=1
            else:
                my_vowel[s[i]]+=1
        else:
            if s[i] not in my_consonants:
                my_consonants[s[i]]=1
            else:
                my_consonants[s[i]]+=1
    a=sorted(my_consonants.values())
    b=sorted(my_vowel.values())
    if len(b)!=0 and len(a)!=0:
        return a[-1]+b[-1]
    if len(b)==0:
        return a[-1]
    if len(a)==0:
        return b[-1]
```

```python
maxFreqSum("aeiaeia")
```
