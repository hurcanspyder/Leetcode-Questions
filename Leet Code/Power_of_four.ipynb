```python
def power(number):
    import math
    log_value=math.log(number,4)
    if log_value==int(log_value):
        return True
    else:
        return False
```

```python
power(16)
```

# Recursion Approach


```python
def power(number):
    if number==4:
    
        return True
    if int(number)!=number:
        return False
    return power(number/4)
    

    
```

```python
power(1)
```
