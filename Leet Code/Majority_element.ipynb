```python
def majority_element1(list1):
    hash_map={}
    n=len(list1)
    for i in range(0,len(list1)):
        if list1[i] not in hash_map:
            hash_map[list1[i]]=1
        else:
            hash_map[list1[i]]+=1
            if hash_map[list1[i]]>=n/2:
                return list1[i]
```
