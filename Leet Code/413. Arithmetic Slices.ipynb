```python
class Solution(object):
    def numberOfArithmeticSlices(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        count=0
        
        for i in range(0,len(nums)-2):
            diff=0
            new=True
            for j in range(i+1,len(nums)):
                if new:
                    diff=nums[j]-nums[i]
                    new=False
                elif nums[j]-nums[j-1]==diff:
                    if j-i>=2:
                        count+=1
                else:
                    break
        return count

        
```

```python
class Solution(object):
    def numberOfArithmeticSlices(self, nums):
        curr = 0
        total = 0
        
        for i in range(2, len(nums)):
            if nums[i] - nums[i-1] == nums[i-1] - nums[i-2]:
                curr += 1
                total += curr
            else:
                curr = 0
        return total
```
