```python
def move_zero(list1):
    list2=[]
    for i in range(0,len(list1)):
        if list1[i]!=0:
            list2.append(list1[i])
    n=len(list1)-len(list2)
   
    list1[:]=list2+[0]*n
    print(list1)

            
```

```python
move_zero([1,0,0,0,2,3,1,0,2])
```

## By Two pointer approach

```python
def shift(list1):
    i=0
    if len(list1)==1:
        return 
    while(i<len(list1)):
        if list1[i]==0:
            break
        i+=1
    if i==len(list1):
        return 
    j=i+1
    while j<len(list1):
        if list1[j]!=0:
            list1[i],list1[j]=list1[j],list1[i]
            i+=1
        j+=1

    print(list1)

            
    
    


```

```python
shift([1,0,0,0,2,3,1,0,2])
```

**Space complexity reduced to O(1) and time complexity is O(n)**
