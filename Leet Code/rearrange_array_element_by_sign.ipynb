```python
def rearrange(list1):
    list2=[]
    list3=[]
    result=[]
    n=len(list1)
    for i in range(0,n):
        if list1[i]<0:
            list2.append(list1[i])
        if list1[i]>=0:
            list3.append(list1[i])
    j=0
    while j<n/2:
        
        result.append(list3[j])
        result.append(list2[j])
        j+=1

        
    return result


```

```python
rearrange([-1,1])
```

# Two Pointer Approach

```python
def rearrange1(list1):
    n=len(list1)
    result=[0]*n
    positive=0
    negative=1
    for i in range(0,n):
        if list1[i]>=0:
            result[positive]=list1[i]
            positive+=2
        if list1[i]<0:
            result[negative]=list1[i]
            negative+=2

    return result
        
```

```python
rearrange1([3,1,-2,-5,2,-4])
```

**Space complexity of this code is O(n) where as the the complexity for the last code was O(n^2) and space complexity here is bit controversial it can be o(1) also and o(n) also all depends on you**
