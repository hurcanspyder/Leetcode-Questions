```python
def Longest_Consecutive_Sequence(list1):
    count,maxii=1,0
    for i in range(0,len(list1)):
        count=1
        num=list1[i]
        while num+1 in list1:
            count+=1
            num=num+1
        maxii=max(maxii,count)
    return maxii
```

```python
Longest_Consecutive_Sequence( [0,3,7,2,5,8,4,6,0,1])
```

# Optimal Approach

```python
def Longest_Consecutive_Sequence1(list1):
    list1.sort()
    count=0
    last_smaller=-10**30
    longest=0
    for i in range(0,len(list1)):
        num=list1[i]
        if num-1==last_smaller:
            count+=1
            last_smaller=num
        elif num!=last_smaller:
            count=1
            last_smaller=num
        longest=max(longest,count)
    return longest
```

```python
Longest_Consecutive_Sequence1( [0,3,7,2,5,8,4,6,0,1])
```
