```python
def binary_search(nums,target):
    n=len(nums)
    low=0
    high=n-1
    while low<=high:
        mid=(low+high)//2
        if nums[mid]==target:
            return mid
        elif nums[mid]<target:
            low=mid+1
        else:
            high=mid-1
    return -1
```

```python
def check(n,count):
    if n//5==0 or n%5!=0:
        return count
    elif n%5==0:
        count+=1
        
    return check(n//5,count)
```
