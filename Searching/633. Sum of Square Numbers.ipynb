```python
import math

class Solution(object):
    def judgeSquareSum(self, c):
        """
        :type c: int
        :rtype: bool
        """
        i = 0
        j = int(math.sqrt(c))  # max possible square root within c

        while i <= j:
            s = i * i + j * j
            if s == c:
                return True
            elif s < c:
                i += 1
            else:
                j -= 1

        return False
```
