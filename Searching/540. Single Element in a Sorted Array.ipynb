```python
class Solution(object):
    def singleNonDuplicate(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        left=0
        right=len(nums)-1
        if len(nums)==1:
            return nums[0]
        while left<=right:
            mid=(left+right)//2
            if mid==len(nums)-1 and nums[mid-1]<nums[mid]:
                return nums[mid]
            if mid==0 and nums[mid]<nums[mid+1]:
                return nums[mid]
            if nums[mid]>nums[mid-1] and nums[mid]<nums[mid+1]:
                return nums[mid]

            if mid%2==0:
                if nums[mid]==nums[mid+1]:
                    left=mid+1
                else:
                    right=mid-1
            else:
                if nums[mid]==nums[mid+1]:
                    right=mid-1
                else:
                    left=mid+1
            

        
```
