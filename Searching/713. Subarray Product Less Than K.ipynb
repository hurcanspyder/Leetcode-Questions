```python
class Solution(object):
    def numSubarrayProductLessThanK(self, nums, k):
        """
        :type nums: List[int]
        :type k: int
        :rtype: int
        """
        l=0
        res=0
        product=1
        for r in range(len(nums)):
            product*=nums[r]
            if product>=k:
                while product>=k and l<=r:
                    product/=nums[l]
                    l+=1
            res+=r-l+1
        return res


        

        
```
