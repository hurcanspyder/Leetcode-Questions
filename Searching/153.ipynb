```python

def findMin(nums):
    low=0
    n=len(nums)
    high=n-1
    min_element=6000
    while low<=high:
        mid=(low+high)//2
        
        if min_element>nums[mid]:
            print("Hello")
            min_element=nums[mid]
        if nums[high]<=nums[mid]:
            low=mid+1
        else:
            high=mid-1
    return min_element
```
