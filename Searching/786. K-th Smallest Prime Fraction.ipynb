```python
class Solution(object):
    def kthSmallestPrimeFraction(self, arr, k):
        """
        :type arr: List[int]
        :type k: int
        :rtype: List[int]
        """
        n = len(arr)
        left, right = 0.0, 1.0

        while True:
            mid = (left + right) / 2.0
            # Count fractions <= mid, and track the maximum one <= mid
            count = 0
            best_num, best_den = 0, 1
            j = 1  # denominator index, non-decreasing through the sweep

            for i in range(n - 1):
                # Move j until arr[i] / arr[j] <= mid  <=>  arr[i] <= mid * arr[j]
                while j < n and arr[i] > mid * arr[j]:
                    j += 1
                if j == n:
                    break  # no denominators left for this i or any larger i
                count += (n - j)
                # The smallest denominator j gives the largest fraction for this i
                if best_num * arr[j] < arr[i] * best_den:
                    best_num, best_den = arr[i], arr[j]

            if count == k:
                return [best_num, best_den]
            elif count < k:
                left = mid
            else:
                right = mid
```

```python
class Solution(object):
    def kthSmallestPrimeFraction(self, arr, k):
        """
        :type arr: List[int]
        :type k: int
        :rtype: List[int]
        """
        fractions = []

        for i in range(len(arr)):
            for j in range(i + 1, len(arr)):
                fractions.append((arr[i] / float(arr[j]), arr[i], arr[j]))

        # Sort by fraction value
        fractions.sort(key=lambda x: x[0])

        # Get the kth smallest fraction (1-indexed)
        _, num, den = fractions[k - 1]

        return [num, den]

        

        
```
