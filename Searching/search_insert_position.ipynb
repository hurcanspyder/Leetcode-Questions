```python
def search(nums,target):
    count=0
    n=len(nums)
    low=0
    high=n-1
    ub=n
    while low<=high:
        mid=(low+high)//2

        if nums[mid]>=target:
            
            ub=mid
            high=mid-1

        else:
            low=mid+1
    return ub
        
    
    
```

```python
search([1,3,5,6],7)
```
