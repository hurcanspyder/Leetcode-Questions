```python
class Solution(object):
    def arrangeCoins(self, n):
        """
        :type n: int
        :rtype: int
        """
        # count=0
        # if n==1:
        #     return 1
        # sum1=0
        # for i in range(1,n):
        #     sum1+=i
        #     if sum1<=n:
        #         count+=1
        #     else:
        #         break
        # return count
        left=1
        right=n
        count=0
        while left<=right:
            mid=(left+right)//2
            if mid*(mid+1)/2<=n:
                count=mid
                left=mid+1
            else:
                right=mid-1
        return count

    
            

        
```
