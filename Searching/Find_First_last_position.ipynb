```python
def find(nums,target):
    n=len(nums)
    low,low1=0,0
    high,high1=n-1,n-1
    index1=[]
    lb=n
    ub=n
    if target not in nums:
        return [-1,-1]
    else:
        while low<=high:
            mid=(low+high)//2
            
            if nums[mid]>=target:

                lb=mid
                high=mid-1
                

  
            else:
                low=mid+1
         
        while low1<=high1:
            mid1=(low1+high1)//2
            if nums[mid1]>target:


                high1=mid1-1
                ub=mid1
            else:
                low1=mid1+1
        return [lb,ub-1]


        

    
```
