```python
class Solution(object):
    def shipWithinDays(self, weights, days):
        """
        :type weights: List[int]
        :type days: int
        :rtype: int
        """
        l=max(weights)
        r=sum(weights)
        def func(shipcapacity):
            capacity=0
            days=1
            for weight in weights:
                if capacity+weight>shipcapacity:
                    days+=1
                    capacity=weight
                else:
                    capacity+=weight
            return days
        while l<r:
            m=(l+r)//2
            if func(m)<=days:
                r=m
            else:
                l=m+1
        return l
```
