Brute force

```python
import math
class Solution(object):
    def smallestDivisor(self, nums, threshold):
        """
        :type nums: List[int]
        :type threshold: int
        :rtype: int
        """
        sum1=sum(nums)
        i=1
        while sum1 >threshold:
            sum1=0
            for j in nums:
                if j<i:
                    sum1+=1
                elif j>i:
                    sum1+=math.ceil(j/float(i))
                else:
                    sum1+=1
            i+=1
        if i==1:
            return 1
        return i-1
```

Optimized

```python
import math

class Solution(object):
    def smallestDivisor(self, nums, threshold):
        left, right = 1, max(nums)
        
        while left < right:
            mid = (left + right) // 2
            total = sum(math.ceil(num / float(mid)) for num in nums)
            
            if total > threshold:
                left = mid + 1
            else:
                right = mid
        
        return left
```
