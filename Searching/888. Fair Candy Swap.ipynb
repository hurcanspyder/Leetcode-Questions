```python
class Solution(object):
    def fairCandySwap(self, aliceSizes, bobSizes):
        """
        :type aliceSizes: List[int]
        :type bobSizes: List[int]
        :rtype: List[int]
        """
        a=sum(aliceSizes)
        b=sum(bobSizes)
        avg=(a+b)/2
        if a<avg:
            for i in aliceSizes:
                if avg-(a-i) in bobSizes:
                    return [i,avg-(a-i)]
        if b<avg:
            for j in bobSizes:
                if avg-(b-j) in aliceSizes:
                    return [avg-(b-j),j]
            



        
```
