```python
def lower_bound(nums,target):
    n=len(nums)
    lb=nums[-1]
    nums.sort()
    low=0
    high=n-1
    for i in range(0,n):
        mid=(low+high)//2
        if nums[mid]>=target:
            lb=mid
            high=mid-1
        else:
            low=mid+1
    return lb
```

```python
print(lower_bound([1,1,1,2,2,2,2,2,2,3,5,6,7,7,8,9,9],4))
```
