```python
def count(nums,target):
    n=len(nums)
    lb=0
    ub=0
    low,low1=0,0
    high,high1=n-1,n-1
    while low<=high:
        mid=(low+high)//2
        if nums[mid]>=target:
            high=mid-1
            lb=mid
        else:
            low=mid+1
    while low1<=high1:
        mid1=(low1+high1)//2
        if nums[mid1]>target:
            high1=mid1-1
            ub=mid1
        else:
            low1=mid1+1
    print(ub-lb)



```

```python
count([1,2,3,3,3,3,3,5,6,8,9,9,10],9)
```
