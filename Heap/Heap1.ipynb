```python
def minimumTime( d, r):
    """
    :type d: List[int]
    :type r: List[int]
    :rtype: int
    """
    i=0
    j=1
    count1=1
    del1=0
    del2=0
    while del1!=d[i] or del2!=d[j]:
        if count1%r[i]!=0 and del1<d[i]:
            del1+=1
        elif count1%r[i]==0 and count1%r[j]!=0 and del2<d[j]:
            del2+=1
        count1+=1 
    return count1
```

```python
def minimumTime1(d, r):
    i, j = 0, 1
    count1 = 1
    del1 = del2 = 0

    while del1 != d[i] or del2 != d[j]:
        if count1 % r[i] != 0 and del1 < d[i]:
            del1 += 1
        elif count1 % r[j] != 0 and del2 < d[j]:
            del2 += 1
        count1 += 1
    count2=0
    del1=0
    del2=0
    while del1 != d[i] or del2 != d[j]:
        if count2 % r[j] != 0 and del2 < d[j]:
            del2 += 1
        elif count2 % r[i] != 0 and del1 < d[i]:
            del1 += 1
        count2 += 1

    return min(count2 - 1,count1-1)
```
