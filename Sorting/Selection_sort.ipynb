```python
def selection_sort(list1):
    for i in range(0,len(list1)):
        min_index=i
        for j in range(i+1,len(list1)):
            if list1[j]<list1[min_index]:
                min_index=j
            temp=list1[i]
            
            list1[i]=list1[min_index]
            list1[min_index]=temp
    print(list1)


```

```python
selection_sort([4,3,2,5,5,1])
```

**Time complexity of this code is O(n^2) and spcae complexity is O(1)**
