```python
def bubble_sort(list1):
    for i in range(len(list1)-2,-1,-1):
        is_swap=False
        for j in range(0,i+1):
            if list1[j]>list1[j+1]:
                list1[j],list1[j+1]=list1[j+1],list1[j]
                is_swap=True
        if is_swap==False:
            break
    print(list1)


```

```python
bubble_sort([4,5,6,4,2,4,6,9,1])
```

**Time complexity of avergae and worst case is O(n^2) and space complexity is O(n) but in best case it becomes O(n) when it is already sorted means no swapping occured till now**
