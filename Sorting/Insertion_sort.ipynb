```python
def insertion_sort(list1):
    
    for i in range(0,len(list1)-1):
        pivot=list1[i]
        if pivot<=list1[i+1]:
            continue
        else:
            pivot=list1[i+1]
            for j in range(i,-1,-1):
                if pivot<list1[j]:
                    temp=list1[j]
                    list1[j]=pivot
                    list1[j+1]=temp
    print(list1)

```

```python
insertion_sort([3,5,6,4,8,9,10,1])
```

**Time complexity for this is also O(n^2) and space complexity is O(1)**
