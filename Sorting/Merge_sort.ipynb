```python
def merge_sort(list1):
    if len(list1) <= 1:
        return list1
    mid = len(list1) // 2
    left = list1[:mid]
    right = list1[mid:]
    left = merge_sort(left)
    right = merge_sort(right)
    return merge_array(left, right)

def merge_array(left, right):
    result = []
    i, j = 0, 0
    n, m = len(left), len(right)
    while i < n and j < m:
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    while i < n:
        result.append(left[i])
        i += 1
    while j < m:
        result.append(right[j])
        j += 1
    return result

# Example usage
print(merge_sort([5, 2, 9, 1, 5, 6]))
```

**The Time complexity of merge sorting is O(log2N) and space complexity is O(n)**
