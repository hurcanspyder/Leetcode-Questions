# Simple approach

```python
def solve(fruits):
    sum1=0
    for i in range(len(fruits)):
        sum=1
        set1=set()
        count=1
        set1.add(fruits[i])
        for j in range(i+1,len(fruits)):
            

            if fruits[j] not in set1:
                set1.add(fruits[j])
                count+=1
            if count>2:
                break
            sum+=1
        sum1=max(sum1,sum)
    return sum1

            

```

```python
solve([3,3,3,1,2,1,1,2,3,3,4])
```

# sliding window Approach

```python
def solve1(fruits):
    left=0
    right=0
    max_length=0
    my_dict={}
    while right<len(fruits):
        if fruits[right] not in my_dict:
            my_dict[fruits[right]]=1
        else:
            my_dict[fruits[right]]+=1
        while len(my_dict)>2:
            my_dict[fruits[left]]-=1
            if my_dict[fruits[left]]==0:
                my_dict.pop(fruits[left])
            left+=1
        
        
        max_length=max(max_length,right-left+1)
        right+=1
    return max_length


        

```

**Sliding window gives us 0(n) approach**

# Optimal approach

```python
def solve2(fruits):
    left=0
    right=0
    max_length=0
    my_dict={}
    while right<len(fruits):
        if fruits[right] not in my_dict:
            my_dict[fruits[right]]=1
        else:
            my_dict[fruits[right]]+=1
        if  len(my_dict)>2:
            my_dict[fruits[left]]-=1
            if my_dict[fruits[left]]==0:
                my_dict.pop(fruits[left])
            left+=1
        
        if len(my_dict)<=2:
            max_length=max(max_length,right-left+1)
        right+=1
    return max_length
```

**I just used last soltuion as a optimal approach just the differnce is instead of using while i used if statment**

```python
solve2([1,2,1])
```
