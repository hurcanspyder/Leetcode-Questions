```python
def minSubArrayLen(target, nums):
    """
    :type target: int
    :type nums: List[int]
    :rtype: int
    """
    window=nums[:2]
    i=0
    j=2
    count=10*10
    while i<len(nums) and j<=len(nums):
        window=nums[i:j]
        print(i,j)
        
        if sum(window)==target:
            print("a")
            count=min(count,len(window))
        if sum(window)>=target:
            print("b")
            i+=1
        if sum(window)<target:  
            j+=1
            print("c")
    return count
```
