```python
def solve(nums,k):
    my_dict={}
    neg=0
    for i in range(len(nums)):
        if nums[i] not in my_dict:
            my_dict[nums[i]]=i
        else:
            neg=abs(my_dict[nums[i]]-i)
            if neg<=k:
                return True
            else:
                my_dict[nums[i]]=i
    return False
            

   
            

```

```python
solve([1,0,1,1],1)
```
