```python
def solve(asteriods):
    stack=[]
    for i in range(0,len(asteriods)):
        if asteriods[i]>0:
            stack.append(asteriods[i])
        else:
            

            while len(stack)!=0 and stack[-1]+asteriods[i]<0 and stack[-1]>0:

                stack.pop()
            if len(stack)!=0 and stack[-1]+asteriods[i]==0:
                stack.pop()
                
            
            elif len(stack)==0 or stack[-1]<0:
                stack.append(asteriods[i])
                   
                    
                
            
            
    return stack
```

```python
solve([8,-8])
```
