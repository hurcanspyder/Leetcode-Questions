# Simple approach

```python
def solve(s):
    max=0
   
    for i in range(len(s)):
        count=1
        set1=set()
        set1.add(s[i])

        for j in range(i+1,len(s)):
            if s[j] not in set1:
                
                count+=1
                set1.add(s[j])
            else:
                break
        if count>max:
            max=count
    return max




```

```python
solve("bbbbb")
```

# optimal approach

```python
def solve(s):
    max=0
   
    for i in range(len(s)):
        count=1
        set1=set()
        set1.add(s[i])

        for j in range(i+1,len(s)):
            if s[j] not in set1:
                
                count+=1
                set1.add(s[j])
            else:
                i=j+1
                break
        if count>max:
            max=count
    return max
```

```python
solve("abcabcbb")
```
