```python
def solve(cardPoints,k):
    if k==len(cardPoints):
        return sum(cardPoints)
    left_sum,right_sum=0,0
    for i in range(0,k):
        left_sum+=cardPoints[i]
    maxii=left_sum
    right_ind=len(cardPoints)-1
    for i in range(k-1,-1,-1):
        left_sum-=cardPoints[i]
        right_sum+=cardPoints[right_ind]
        maxii=max(maxii,left_sum+right_sum)
        right_ind-=1
    return maxii
        
```

```python
solve([2,2,2],2)
```
