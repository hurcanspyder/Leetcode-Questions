```python
class Solution(object):
    def findLongestWord(self, s, dictionary):
        """
        :type s: str
        :type dictionary: List[str]
        :rtype: str
        """
        dictionary.sort(key=lambda x: (-len(x), x))
        # for i in range(len(dictionary)):
        maxi=0
        str1=""
        for i in range(len(dictionary)):
            left=0
            right=0
            while  right<=len(s):
                if left==len(dictionary[i]):
                    if len(dictionary[i])>maxi:
                        str1=dictionary[i]
                        maxi=len(dictionary[i])
                        break
                    break
                if right==len(s):
                    break

                    
                if dictionary[i][left]==s[right]:
                    left+=1
                    right+=1
                else:
                    right+=1
        return str1
        
```
