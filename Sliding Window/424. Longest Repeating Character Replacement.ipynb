```python
class Solution(object):
    def characterReplacement(self, s, k):
        """
        :type s: str
        :type k: int
        :rtype: int
        """
    
        left=0
        right=1
        count1=0
        maxi=0
        while left<=right and right<len(s):
            if s[left]==s[right]:
                
                maxi=max(maxi,right-left+1)
                right+=1
            elif s[left]!=s[right]:
                if count1<k:
                    
                    maxi=max(maxi,right-left+1)
                    right+=1
                    count1+=1
                else:
                    left+=1
                    maxi=max(maxi,right-left+1)
        return maxi

                    
                    
            
            

        
```
