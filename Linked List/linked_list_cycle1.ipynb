```python
class Solution(object):
    def hasCycle(self, head):
        """
        :type head: ListNode
        :rtype: bool
        """
        set1=set()

        temp=head
        while temp is not None:
            if temp.next in set1:
                return True 
            else:
                temp=temp.next
                set1.add(temp)

        return False
```

# Floyd's Cycle Detection

```python
class Solution(object):
    def reverseList(self, head):
        """
        :type head: Optional[ListNode]
        :rtype: Optional[ListNode]
        """
        slow=head
        fast=head
        while fast is not None and fast.next is not None:
            slow=slow.next
            fast=fast.next.next

            if slow==fast:
                return True
        return False
```

**Space complexity os Floyd's detection is O(1) where as normally it si O(n)**
