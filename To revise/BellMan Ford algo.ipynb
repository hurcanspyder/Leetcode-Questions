# Single Source Shortest Path

1.Dijkstra's Algo:- Can only work when there is no negative edges

2.Bellman Ford Algo:- Can work in any condition where it is negative cycle or negative weights are present(*Cause this algo works on relaxation where it relases the edges v-1 times and if after v-1 times we will recheck and if distance array get changed means particular graph contain negative cycle*)

```python

def func(edges,v,src):
    dis=[10**8]*v
    dis[src]=0
    for i in range(v-1):
        for u,v,w in edges:
            dis[v]=min(dis[v],dis[u]+w)

    for u,v,w in edges:
        if dis[v]>dis[u]+w:
            print("Negative weight cycle detect")
    return dis
```
