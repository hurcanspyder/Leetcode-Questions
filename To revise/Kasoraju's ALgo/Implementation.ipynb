# Steps
1. Apply a simple DFS and reverse it in stack 
2. Reverese the whole graph 
3. Now apply DFS again but pick the elements from the stack and store them in one list in one go and mark visited as True so that we dont tarverse them again now the differnt list we will get is SCC.

🔥 Complete Flow (Very Important)

Run DFS → fill stack

Reverse graph

Pop from stack:

If not visited → DFS → get SCC

🔥 Key Insight (Golden Line)

“Stack stores nodes in decreasing finishing time — that’s why SCCs come out correctly”

```python
def Kasorajus(adj_list,n):
    def dfs1(n,visited,st,adj_list):
        print(n,end="")
        visited[n]=True
        for adj_node in adj_list[n]:
            if not visited[adj_node]:
                dfs1(adj_node,visited,st,adj_list)
        st.append(n)
    
    
    
    def dfs2(visited,scc,n,adj_list2):
        visited[n]=True
        scc.append(n)
        for adj_node in adj_list2[n]:
            if not visited[adj_node]:
                dfs2(visited,scc,n,adj_list2)
    
    st=[]
    visited=[False]*(n+1)
    adj_list1=[[] for _ in range(n+1)]
    for node in range(n):
        for i in adj_list[node]:
            adj_list1[i].append(node)
    for i in range(n):
        if not visited[i]:
            dfs1(i,visited,st,adj_list)
    
    visited1=[False]*(n+1)

    SCC=[]

    while st:
        u=st.pop()
        if not visited1[u]:
            scc=[]
            dfs2(visited1,scc,u,adj_list1)
            SCC.append(scc)
    return SCC


    
       

    
    


```
