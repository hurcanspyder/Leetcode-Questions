# Suffix-Prefix solution

```python
class Solution(object):
    def maxProduct(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        n=len(nums)
        prefix=1
        suffix=1
        maxi=float('-inf')
        for i in range(0,n):
            prefix*=nums[i]
            suffix*=nums[n-i-1]
            maxi=max(maxi,suffix,prefix)
            if prefix==0:
                prefix=1
            if suffix==0:
                suffix=1
        return maxi
```
