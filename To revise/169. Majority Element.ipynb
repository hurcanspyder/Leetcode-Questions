*Actually in This code we are keeping the track if maximum element which is present in the array for eg we have one array[2,2,1,1,1,2,2] now in this array majoriy element is 2 which means we have to return this number but in   TC->O(N) AND S(C)->O(1) so for this we will keep track of first number and we will reduce the count after every different digit and increase after same digit and when count become 0 we will assign candidate new digit* 

```python
class Solution:
    def majorityElement(self, nums):
        count = 0
        candidate = 0
        
        for num in nums:
            if count == 0:
                candidate = num
            
            if num == candidate:
                count += 1
            else:
                count -= 1
        
        return candidate
```
