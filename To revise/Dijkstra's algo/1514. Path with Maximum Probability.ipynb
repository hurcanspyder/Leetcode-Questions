```python
class Solution(object):
    def maxProbability(self, n, edges, succProb, start_node, end_node):
        """
        :type n: int
        :type edges: List[List[int]]
        :type succProb: List[float]
        :type start_node: int
        :type end_node: int
        :rtype: float
        """
        adj_list=[[] for _ in range(n)]
        i=0
        for u,v in edges:
            adj_list[u].append([v,succProb[i]])
            adj_list[v].append([u,succProb[i]])
            i+=1
        


        src=start_node
        from collections import deque
        queue=deque()
        queue.append([src,1])
        distance=[-10**7]*n
        prb=[0]*n
        distance[src]=1
        while queue:
            e,prob=queue.popleft()
            for adj_node,adj_prob in adj_list[e]:
                if distance[adj_node]<prob*adj_prob:
                    distance[adj_node]=max(distance[adj_node],prob*adj_prob)
                    
                    queue.append([adj_node,distance[adj_node]])
        if distance[end_node]!=-10**7:
            return distance[end_node]
        return 0.00000
```
