# Brute force approach

```python
class Solution(object):
    def nthUglyNumber(self, n, a, b, c):
        """
        :type n: int
        :type a: int
        :type b: int
        :type c: int
        :rtype: int
        """
        ia,ib,ic=1,1,1
        list1=[]
        for i in range(1,n+1):
            z=min(ia*a,ib*b,ic*c)
            list1.append(z)
            if z==ia*a:
                ia+=1
            if z==ib*b:
                ib+=1
            if z==ic*c:
                ic+=1
        # print(list1)
        return list1[-1]

```
