```python
class Solution(object):
    def combine(self, n, k):
        """
        :type n: int
        :type k: int
        :rtype: List[List[int]]
        """
        nums=[i for i in range(1,n+1)]
        def func(index,subset,result,k,nums):
            if len(subset)==k:
                result.append(subset[:])
                return
            if index>=len(nums):
                return 
            
            subset.append(nums[index])
            func(index+1,subset,result,k,nums)
            subset.pop()
            func(index+1,subset,result,k,nums)
            return result
        return func(0,[],[],k,nums)

    
        
```
