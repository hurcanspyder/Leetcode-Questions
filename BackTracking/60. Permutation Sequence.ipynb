# optimal way

```python
class Solution:
    def getPermutation(self, n, k):
        import math
        digits = list(range(1, n + 1))
        result = []
        k -= 1
        
        for i in range(n, 0, -1):
            fact = math.factorial(i - 1)
            idx = k // fact
            result.append(str(digits[idx]))
            digits.pop(idx)
            k %= fact
        
        return ''.join(result)
```
