```python
class Solution(object):
    def letterCasePermutation(self, s):
        """
        :type s: str
        :rtype: List[str]
        """
       
        
        def func(index, s, subset, result):
            if index == len(s):
                result.append("".join(subset))
                return

            if s[index].isdigit():
                subset.append(s[index])
                func(index + 1, s, subset, result)
                subset.pop()
                return

            subset.append(s[index].upper())
            func(index + 1, s, subset, result)
            subset.pop()

            subset.append(s[index].lower())
            func(index + 1, s, subset, result)
            subset.pop()


        result = []
        func(0, s, [], result)
        return result

        








        
```
