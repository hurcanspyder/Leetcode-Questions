```python
class Solution(object):
    def numberOfBoomerangs(self, points):
        """
        :type points: List[List[int]]
        :rtype: int
        """
        count=0
        for x,y in points:
            counter={}
            for a,b in points:
                key=(a-x)**2+(b-y)**2
                if key in counter:
                    count+=2*counter[key]
                    counter[key]+=1
                else:
                    counter[key]=1
        return count
```
