```python
class Solution(object):
    def searchMatrix(self, matrix, target):
        """
        :type matrix: List[List[int]]
        :type target: int
        :rtype: bool
        """
        left=0
        right=len(matrix[0])
        up=0
        down=len(matrix)
        while left<right and up<down:
            mid_x=(left+right)//2
            mid_y=(up+down)//2
            if matrix[mid_y][mid_x]==target:
                return True
            if matrix[mid_y][mid_x]>target:
                if matrix[mid_y][0]>target:
                    down=mid_y

                else:
                    right=mid_x
            if matrix[mid_y][mid_x]<target:
                if matrix[mid_y][-1]<target:
                    up=mid_y+1
                else:
                    left=mid_x+1
        return False

                


        
```
