```python
class Solution(object):
    def minMergeCost(self, lists):
        """
        :type lists: List[List[int]]
        :rtype: int
        """

        def func(used, subset, total, subset1, result):
            if len(subset) == len(lists):
                result.append(total)
                return

            for index in range(len(lists)):
                if not used[index]:
                    used[index] = True
                    subset.append(lists[index])

                    sum1 = total
                    if subset1:
                        if len(subset1) % 2 != 0 and len(lists[index]) % 2 != 0:
                            sum1 += len(subset1) + len(lists[index]) + abs(
                                subset1[len(subset1)//2] - lists[index][len(lists[index])//2]
                            )
                        elif len(subset1) % 2 == 0 and len(lists[index]) % 2 != 0:
                            sum1 += len(subset1) + len(lists[index]) + abs(
                                subset1[len(subset1)//2 - 1] - lists[index][len(lists[index])//2]
                            )
                        elif len(subset1) % 2 == 0 and len(lists[index]) % 2 == 0:
                            sum1 += len(subset1) + len(lists[index]) + abs(
                                subset1[len(subset1)//2 - 1] - lists[index][len(lists[index])//2 - 1]
                            )
                        else:
                            sum1 += len(subset1) + len(lists[index]) + abs(
                                subset1[len(subset1)//2] - lists[index][len(lists[index])//2 - 1]
                            )

                    old_subset1 = subset1[:]              # 🔥 BACKUP
                    subset1 = sorted(subset1 + lists[index])

                    func(used, subset, sum1, subset1, result)

                    subset1 = old_subset1                  # 🔥 RESTORE
                    subset.pop()
                    used[index] = False

        result = []
        func([False]*len(lists), [], 0, [], result)
        
        return min(result)

                
            
        
        ©leetcode
```
