```python
class Solution(object):
    def wordBreak(self, s, wordDict):
        """
        :type s: str
        :type wordDict: List[str]
        :rtype: bool
        """
        def func(subset,wordDict,s):
            if "".join(subset)==s:
                return True
            if len("".join(subset))>=len(s) and "".join(subset)!=s:
                return
                

            
            for word in wordDict:
                subset.append(word)
                if func(subset,wordDict,s):
                    return True
                subset.pop()
            return False
        return func([],wordDict,s)


        
```

```python
class Solution(object):
    def wordBreak(self, s, wordDict):
        """
        :type s: str
        :type wordDict: Set[str]
        :rtype: bool
        """
        dp = [False] * (len(s) + 1) # dp[i] means s[:i+1] can be segmented into words in the wordDicts 
        dp[0] = True
        for i in range(len(s)):
            for j in range(i, len(s)):
                if dp[i] and s[i: j+1] in wordDict:
                    dp[j+1] = True
                    
        return dp[-1]
```
