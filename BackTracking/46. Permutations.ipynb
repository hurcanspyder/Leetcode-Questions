```python
class Solution(object):
    def permute(self, nums):
        """
        :type nums: List[int]
        :rtype: List[List[int]]
        """
        def func(subset,nums,used,result):
            if len(subset)==len(nums):
                result.append(subset[:])
                return 
            
            for i in range(len(nums)):
                if not used[i]:
                    subset.append(nums[i])
                    used[i]=True
                    func(subset,nums,used,result)
                    subset.pop()
                    used[i]=False
        result=[]
        func([],nums,[False]*len(nums),result)
        return result
```
