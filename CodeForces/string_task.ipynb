```python
def string_task(string1):
    vowels = "aeiouyAEIOUY"
    list1 = ['']
    for i in string1:
        if i not in vowels and i.isalpha():  # Remove vowels, keep only letters
            if 65<=ord(i)<=90:
                j=chr(ord(i)+32)
                list1.append(j)
            else:
                list1.append(i)


    string1 = '.'.join(list1)
    print(string1)

string1=str(input())
string_task(string1)

            
```

```python
string_task("tour")
```
