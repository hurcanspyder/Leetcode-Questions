```python
def registeration_system(number, names_list):
    name_count = dict()
    result = []

    for name in names_list:
        if name not in name_count:
            name_count[name] = 1
            result.append("OK")
        else:
            result.append(name + str(name_count[name]))
            name_count[name] += 1

    for res in result:
        print(res)


# Input handling
number = int(input())
names_list = []

for _ in range(number):
    name = input().strip()
    names_list.append(name)

registeration_system(number, names_list)
```
