```python
def theater_square(list1):
    area_of_rectangle=list1[0]*list1[1]
    area_of_square=list1[2]*list1[2]
    side_of_rectangle1=list1[0]
    side_of_rectangle2=list1[1]
    side_of_square=list1[2]
    if area_of_square<area_of_rectangle:
        count=0
        while(side_of_rectangle1>0):
            if side_of_rectangle1==side_of_square:
                break
            else:
                side_of_rectangle1=side_of_rectangle1-side_of_square
                count=count+1
        while(side_of_rectangle2>0):
            if side_of_rectangle2==side_of_square:
                break
            else:
                side_of_rectangle2=side_of_rectangle2-side_of_square
                count=count+1
    else:
        count=1
    print(count)

list1 = list(map(int, input().split()))
theater_square(list1)     
```
