```python
def iq_test(number,list1):
    a=0
    b=0
    list2=[]
    list3=[]
    for i in range(0,number):
        if list1[i]%2==0:
            list2.append(i+1)
            a=a+1
            
        else:
            list3.append(i+1)
            b=b+1
    if a==1:
        print(list2[0])
    if b==1:
        print(list3[0])
number=int(input())
list1 = list(map(int, input().split()))
iq_test(number,list1)
        

```
