```python
def max_points(n, nums):
    from collections import Counter

    count = Counter(nums)
    max_num = max(nums)

    dp = [0] * (max_num + 2)  # To avoid index error at dp[x-2]

    dp[1] = count[1] * 1

    for x in range(2, max_num + 1):
        dp[x] = max(dp[x - 1], dp[x - 2] + count[x] * x)

    return dp[max_num]
```
