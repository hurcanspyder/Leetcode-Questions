```python
def cheap_travel(n, m, a, b):
    # Option 1: All single rides
    cost1 = n * a

    # Option 2: m-ride + single tickets
    cost2 = (n // m) * b + (n % m) * a

    # Option 3: Only m-ride tickets, possibly extra rides
    cost3 = ((n + m - 1) // m) * b

    print(min(cost1, cost2, cost3))

# Input
n, m, a, b = map(int, input().split())
cheap_travel(n, m, a, b)


```
