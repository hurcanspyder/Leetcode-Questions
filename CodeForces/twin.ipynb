```python
def twin(number, list1):
    total = sum(list1)
    half = total / 2

    list2 = sorted(list1, reverse=True)
    count = 0
    curr_sum = 0
    for i in list2:
        curr_sum += i
        count += 1
        if curr_sum > half:
            print(count)
            break

number = int(input())
list1 = list(map(int, input().split()))
twin(number, list1)
```
