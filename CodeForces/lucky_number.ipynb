```python


def is_almost_lucky(number):
    def is_lucky(number):
        for digit in str(number):
            if digit != '4' and digit != '7':
                return False
        return True
    # Generate all lucky numbers up to n
    for i in range(1, number + 1):
        if is_lucky(i) and number % i == 0:
            return "YES"
    return "NO"

# Example usage
n = int(input("Enter a number: "))
print(is_almost_lucky(n))
```
