```python
def ultra_fast(number1,number2):
    number3=[]
    for i in range(0,len(number1)):
        if number1[i]==number2[i]:
            number3.append(0)
        else:
            number3.append(1)
    number3=str("".join(map(str,number3)))
    print(number3)
number1=str(input())
number2=str(input())
ultra_fast(number1,number2)

```

```python
ultra_fast(1010100,1100101)
```
