```python
class Solution(object):
    def minAddToMakeValid(self, s):
        """
        :type s: str
        :rtype: int
        """
        count=0
        stack=[]
        for i in s:
            if i=="(":
                stack.append(i)
            
            else:
                if stack:
                    stack.pop()
                else:
                    count+=1
        return count+len(stack)
            
        
```
