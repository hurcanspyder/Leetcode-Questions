```python
def solve(nums):
    n=len(nums)
    stack=[]
    ans=[-1]*n
    for i in range(len(nums)-1,-1,-1):
        while  len(stack)!=0 and nums[i]>=stack[-1]:
            stack.pop()
        if len(stack)!=0:
            ans[i]=stack[-1]
        stack.append(nums[i])
    return ans
        
        

```

```python
solve([19,2,4,9,3,5,8,10])
```
