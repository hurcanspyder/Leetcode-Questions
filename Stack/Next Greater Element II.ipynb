# Monotnic stack

```python
def solve(nums):
    n=len(nums)
    nums=nums+nums
    a=len(nums)
    result=[-1]*a
    stack=[]
    for i in range(len(nums)-1,-1,-1):
        while len(stack)!=0 and nums[i]>=stack[-1]:
            stack.pop()
        if len(stack)!=0:
            result[i]=stack[-1]
        stack.append(nums[i])
    return result[0:n]

    
    
            

        
        
```

```python
solve([1,2,3,4,3])
```
