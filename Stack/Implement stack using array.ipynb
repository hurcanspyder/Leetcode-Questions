```python
class stack:
    def __init__(self):
        self.items = []
    
    def is_empty(self):
        return self.items == []
    
    def push(self, item):
        self.items.append(item)
    
    def pop(self):
        if len(self.items)==0:
            return "Cannot pop , stack is empty"
        return self.items.pop()

    def top(self):
        if len(self.items)==0:
            return "Stack is empty"
        return self.items[-1]
    
    def size(self):
        return len(self.items)
    
```
