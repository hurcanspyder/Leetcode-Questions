```python
def print_1_to_n(i,n):
    if i>n:
        return 
    print(i)
    print_1_to_n(i+1,n)
```

```python
print_1_to_n(5,9)
```

# Back Tracking

**ALways make recursion trees whenever working with recursion it is also known as back tracking because we are tracking one valaue here**


```python
def print_n_to_1(i,n):
    if i>n:
        return 
    
    print_1_to_n(i+1,n)
    print(i)
    
```

```python
print_n_to_1(5,9)
```

**The only differnce between first and second code is that in second code we are writting print(i) below the function which will print number when the recursion stop level get achieve other wise it will not print anything and when the our thershold will get achieve then the number will get start printing from opposite manner also known as BACK TRACKING**
