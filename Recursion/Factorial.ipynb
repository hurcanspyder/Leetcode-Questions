```python
def factorial(mul,n):
    if n<1:
        print(mul)
        return 
    factorial(mul*n,n-1)
    
```

```python
factorial(1,6)
```

# Or

```python
def factorial1(num):
    if num==0 or num==1:
        return 1
    return num*factorial1(num-1)
```

```python
factorial1(6)
```

**Here time complexity is O(n) and spacec complexity is also O(n) because it is a stackspace**
