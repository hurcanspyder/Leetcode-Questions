```python
def fibonacci_number(first,second,n,count):
    if n==0:
        print(0)
    if n==1:
        print(1)
    elif count==n:
        print(first+second)
        return 
    fibonacci_number(second,first+second,n,count+1)

```

```python
fibonacci_number(0,1,0,1 )
```

**Time complexity of this code is O(2^n) and space is also same**

```python
def fibonacci_number1(num):
    if num==0 or num==1:
        return num
    return fibonacci_number1(num-1)+fibonacci_number1(num-2)
```

```python
fibonacci_number1(6)
```
