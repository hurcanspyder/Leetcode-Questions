```python
def reverse_array(n,list1):
    if n==0:
        return 
    print(list1[n-1])
    reverse_array(n-1,list1)


    
```

```python
reverse_array(10,[1,2,3,4,5,6,7,8,9,0])
```

```python
def reverse_array1(n,list1,list2):
    if n==0:
        print(list2)
    list2.append(list1[n-1])
    reverse_array(n-1,list1)
```

```python
reverse_array1(4,[1,2,3,4],[])
```

**What if we want to reverse the particular sub list in the array**

```python
def reverse_sub_array(start,end,list1):
    if end<=start:
        print(list1)
        return
    a=list1[start-1]
    list1[start-1]=list1[end-1]
    list1[end-1]=a
    reverse_sub_array(start+1,end-1,list1)

   
        
```

```python
reverse_sub_array(1,6,[1,2,3,4,5,6])
```

**Best code one of all The time complexity is O(n/2) and space complexity is O(n)**

# Using while Loop

```python
def reverse(start,end,list1):
    while start<end:
        list1[start-1],list1[end-1]=list1[end-1],list1[start-1]
        start+=1
        end-=1
    print(list1)
```

```python
reverse(2,5,[1,2,3,4,5,6])
```
