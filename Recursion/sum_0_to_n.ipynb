# Using Parameterized recursion function

```python
def sum_0_to_n(sum,i,n):
    if i>n:
        print(sum)
        return
    sum_0_to_n(sum+i,i+1,n)
```

```python
sum_0_to_n(0,1,10)
```

# Using Fucntional recursion

**The only differnce between them is that in parameterixzed recursion we are not returing anything we are just printing but in case of functional recursion we have to return the result which is desired**

1) Create a flow
2) Create a base condition

which in our case is like:

flow: F(n)=N+F(N-1)

Base Conditon: F(1)

```python
def func(N):
    if N==1:
        return 1
    return N+func(N-1)
```

```python
func(10)
```
