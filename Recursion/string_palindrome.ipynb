# By two pointer approach without recursion

```python
def string_palindrome1(string1,left,right):
    if (left>=right):
        print("YES")
        return
    elif(string1[left]==string1[right]):
        string_palindrome1(string1,left+1,left-1)
    else:
        print("NO")
        return
```

```python
string_palindrome1('nitind',0,5)
```
