# Fibonacci?

# Recursion(top-down approach)

```python
def fib1(num):
    if num==0:
        return 0
    if num==1:
        return 1
    return fib1(num-1)+fib1(num-2)
```

TC->O(2**n) SC->O(n)

# Memoization

*Ye tb hoga jb isme sub problems bn rahe honge agr memoization ho para hoga to tabulation bhi ho jaaega*

```python
def fib2(num,dp):
    if num==0:
        return 0
    if num==1:
        return 1
    if dp[num]!=-1:
        return dp[num]
    dp[num]=fib2(num-1,dp)+fib2(num-2,dp)
    return dp[num]
```

TC->O(n) and SC->o(n) (stack space)+O(n) (DP array)

# Tabulation (Bottom-up approach)

```python
def fib3(num):
    dp=[-1]*(num+1)
    dp[0],dp[1]=0,1
    for num in range(2,len(dp)):
        dp[num]=dp[num-1]+dp[num-2]
    return dp[len(dp)-1]
```

TC->O(n) and SC->O(n) Dp array 

# Tabulation (Space Optimization)

*Ye tb hi ho paaega jb piche ki puri array use nahi hori hogi like in our case we are using just last two digits of an array to find out the post one*

```python
def fib4(num):
    prev1=0
    prev2=1
    curr=0
    for i in range(2,num+1):
        curr=prev1+prev2
        prev1=prev2
        prev2=curr
    return curr
```

TC->O(n) and SC->O(1) Just three extra variables we have taken
