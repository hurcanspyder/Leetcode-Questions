```python
class Solution(object):
    def numTeams(self, rating):
        """
        :type rating: List[int]
        :rtype: int
        """
        n=len(rating)
        upper_dp=[0]*n
        lower_dp=[0]*n
        count=0
        for i in range(n):
            for j in range(i):
                if rating[j]<rating[i]:
                    count+=upper_dp[j]
                    upper_dp[i]+=1
                else:
                    count+=lower_dp[j]
                    lower_dp[i]+=1
        return count



            
        
```
