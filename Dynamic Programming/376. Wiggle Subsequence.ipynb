# Tabulation

```python
class Solution(object):
    def wiggleMaxLength(self, nums):
        if len(nums) < 2:
            return len(nums)

        diff = []
        for i in range(1, len(nums)):
            diff.append(nums[i] - nums[i-1])

        # special case
        if all(d == 0 for d in diff):
            return 1   # only equal values

        n = len(diff)
        dp = [1] * n

        for i in range(n):
            for j in range(i):
                if diff[i] > 0 and diff[j] < 0:
                    dp[i] = max(dp[i], dp[j] + 1)

                if diff[i] < 0 and diff[j] > 0:
                    dp[i] = max(dp[i], dp[j] + 1)

        return max(dp) + 1


        

                    

        
        
```

TC->O(n**2)

# greedy approach

```python
class Solution(object):
    def wiggleMaxLength(self, nums):
        if len(nums) < 2:
            return len(nums)

        prev = 0
        count = 1

        for i in range(1, len(nums)):
            curr = nums[i] - nums[i-1]
            if (curr > 0 and prev <= 0) or (curr < 0 and prev >= 0):
                count += 1
                prev = curr

        return count
```

TC->O(n)
