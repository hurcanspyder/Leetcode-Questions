```python
def lcsstr(text1, text2):
    maxi = [0]

    def solve(i, j, count):
        if i == len(text1) or j == len(text2):
            maxi[0] = max(maxi[0], count)
            return

        if text1[i] == text2[j]:
            solve(i+1, j+1, count+1)
        else:
            maxi[0] = max(maxi[0], count)
            solve(i+1, j, 0)
            solve(i, j+1, 0)

    solve(0, 0, 0)
    return maxi[0]

    
```

```python
def tabulation(text1,text2):
    m=len(text1)
    n=len(text2)
    maxi=0
    x=0
    y=0
    dp=[[0 for _ in range(n+1)]for _ in range(m+1)]
    for i in range(1,m+1):
        for j in range(1,n+1):
            if text1[i-1]==text2[j-1]:
                dp[i][j]=1+dp[i-1][j-1]
                if maxi<dp[i][j]:
                    x=i
                    y=j

               

            else:
                dp[i][j]=0
    return [x,y]
    # Now return max out of the dp if not return 0
```
