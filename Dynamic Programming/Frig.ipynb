```python
heights = [10, 5, 20, 0, 15] 
k = 2
def func(index, heights, k,dp):
    if index == len(heights) - 1:
        return 0

    mini = 10**8
    if dp[index]!=-1:
        return dp[index]

    for i in range(1, k+1):
        if index + i < len(heights):
            cost = abs(heights[index+i] - heights[index]) + func(index+i, heights, k,dp)
            mini = min(mini, cost)


    return mini
dp=[-1 for _ in range(len(heights))]
func(0,heights,k,dp)

    

```
