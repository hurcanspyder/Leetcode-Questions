```python
def func(index,list1,k,subset,total):
    if total==k:
        return True
    if total>k:
        return False
    if index>=len(list1):
        return False
    
    subset.append(list1[index])
    sum=total+list1[index]
    pick=func(index+1,list1,k,subset,sum)
    if pick==True:
        return True
    subset.pop()
    not_pick=func(index+1,list1,k,subset,total)
    return not_pick
```

```python
func(0,[1,2,3,45,6],46,[],0)
```

```python
def func1(index,nums,total):
    if total==0:
        return True
    if index==0:
        if nums[0]==total:
            return True
        else:
            return False
       
    if nums[index]>total:
        pick =False
    else:

        pick=func1(index-1,nums,total-nums[index])
    not_pick=func1(index-1,nums,total)
    return not_pick or pick
```

# Memoization

```python
def func1(index,nums,total,dp):
    if total==0:
        return True
    if index==0:
        if nums[0]==total:
            return True
        else:
            return False
    if dp[index][total]!=False:
        return dp[index][total]
       
    if nums[index]>total:
        pick =False
    else:

        pick=func1(index-1,nums,total-nums[index],dp)
    not_pick=func1(index-1,nums,total,dp)
    dp[index][total]= not_pick or pick
    return dp[index][total]
nums=[1,2,3,45,6]
Target=46
dp=[[False for _ in range(Target+1)]for _ in range(len(nums))]
func1(len(nums)-1,nums,Target,dp)
```

SC->O(2len(nums))

# Tabulation

```python
def func2(index,nums,target,dp):
    for i in range(0,len(nums)):
        dp[i][0]=True
    if nums[0]<=target:
        dp[0][nums[0]]==True
    for i in range(1,len(nums)):
        for t in range(0,target+1):
            if nums[i]>t:
                pick=False
            else:
                pick=dp[i-1][t-nums[i]]
            not_pick=dp[i-1][t]
            dp[i][t]=pick or not_pick
    return dp[len(nums)-1][target]
```

*Here the dp contain whether for any i,j before m,n conatin True ? if yes then return True*
