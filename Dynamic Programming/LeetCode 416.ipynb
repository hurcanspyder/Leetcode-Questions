*The question is just like finding out that the subset with sum equivalent to half of the whole nums is present in thw array or not if not then just return False if yes then print return true*

```python
class Solution(object):
    def canPartition(self, nums):
        """
        :type nums: List[int]
        :rtype: bool
        """
        if sum(nums)%2!=0:
            return False
        
        def func(index,nums,total,dp):
            if total==0:
                return True
            if index==0:
                if nums[0]==total:
                    return True
                return False
            if dp[index][total]!=-1:
                return dp[index][total]
            if nums[index]>total:
                pick=False
            else:

                pick=func(index-1,nums,total-nums[index],dp)
            not_pick=func(index-1,nums,total,dp)
            dp[index][total]=pick or not_pick
            return dp[index][total]
        dp=[[-1 for _ in range(sum(nums)/2+1)]for _ in range(len(nums))]
        return func(len(nums)-1,nums,sum(nums)/2,dp)
```
