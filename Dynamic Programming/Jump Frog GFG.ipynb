```python
def jump(index,height):
    if index==0:
        return 0
    
    jump1=jump(index-1)+abs(height[index]-height[index-1])
    if index>1:

        jump2=jump(index-2)+abs(height[index]-height[index-1])
    else:
        jump2=10**8
    return min(jump1,jump2)
```

```python
def jump2(index,height,dp):
    if index==0:
        return 0
    if dp[index]!=-1:
        return dp[index]
    
    jum1=jump2(index-1)+abs(height[index]-height[index-1])
    if index>1:

        jum2=jump2(index-2)+abs(height[index]-height[index-1])
    else:
        jum2=10**8
    dp[index]=min(jum1,jum2)
```

```python
def jump3(index,height,dp):
    dp[0]=0
    for i in range(1,len(dp)):
        jum1=dp[i-1]+abs(height[i]-height[i-1])
        if i>1:
            jum2=dp[i-2]+abs(height[i]-height[i-2])
        else:
            jum2=10**8
        dp[i]=min(jum1,jum2)
    return dp[-1]
```
