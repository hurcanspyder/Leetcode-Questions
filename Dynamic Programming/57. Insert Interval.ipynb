```python
class Solution(object):
    def insert(self, intervals, newInterval):
        """
        :type intervals: List[List[int]]
        :type newInterval: List[int]
        :rtype: List[List[int]]
        """
        stack=[]
        list1=[]
        intervals.append(newInterval)
        intervals.sort()
        for start,end in intervals:
            if len(stack)==0:
                stack.append([start,end])
            else:
                if (start>=stack[0][0] and start<=stack[0][1]) or (stack[0][0]>=start and stack[0][0]<=end):
                    stack[0]=[min(stack[0]+[start,end]),max(stack[0]+[start,end])]
                elif (end>=stack[0][0] and end<=stack[0][1]) or (stack[0][1]>=start and stack[0][1]<=end):
                    stack[0]=[min(stack[0]+[start,end]),max(stack[0]+[start,end])]
                else:
                    e=stack.pop()
                    list1.append(e)
                    stack.append([start,end])
                print(stack)
        if stack:
            e=stack.pop()
            list1.append(e)
        return list1
        
```
