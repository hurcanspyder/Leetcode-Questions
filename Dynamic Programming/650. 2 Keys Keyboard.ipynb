```python
class Solution(object):
    def minSteps(self, n):
        """
        :type n: int
        :rtype: int
        """
        if n==1:
            return 0
        dp=[[-1 for _ in range(n+1)]for _ in range(n+1)]
        def func(index,last_copy,n,dp):
            if index==n:
                return 1
            if index>n:
                return 10**8
            if dp[index][last_copy]!=-1:
                return dp[index][last_copy]

            pick=1+func(index+last_copy,last_copy,n,dp)
            
            not_pick=2+func(2*index,index,n,dp)
            dp[index][last_copy]=min(pick,not_pick)
            return dp[index][last_copy]
        return func(1,1,n,dp)
```
