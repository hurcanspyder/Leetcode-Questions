# we have taken offset cause we are dealing with negative number also here which leads to index error so to manage it we will use offset

# Right One

```python
class Solution(object):
    def numberOfWays(self, startPos, endPos, k):
        MOD = 10**9 + 7
        
        offset = k  # to handle negative positions safely
        init=startPos+2*k+1
        dp = [[-1 for _ in range(k + 1)] for _ in range(init)]
        
        def func(start, end, count):
            # base case
            if count == k:
                return 1 if start == end else 0
            
            # dp lookup
            if dp[start + offset][count] != -1:
                return dp[start + offset][count]
            
            # recursive calls
            right = func(start + 1, end, count + 1)
            left  = func(start - 1, end, count + 1)
            
            dp[start + offset][count] = (left + right) % MOD
            return dp[start + offset][count]
        
        return func(startPos, endPos, 0)
```
