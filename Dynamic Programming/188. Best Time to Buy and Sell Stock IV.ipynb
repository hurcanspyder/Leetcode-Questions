```python
class Solution(object):
    def maxProfit(self, k, prices):
        """
        :type k: int
        :type prices: List[int]
        :rtype: int
        """
        n=len(prices)
        dp=[[[-1 for _ in range(k+1)]for _ in range(2)]for _ in range(n)]
        def func(index,buy,prices,cap):
            if index==len(prices) or cap==0:
                return 0
            if dp[index][buy][cap]!=-1:
                return dp[index][buy][cap]
            if buy==1:
                buy_option=-prices[index]+func(index+1,0,prices,cap)
                not_buy=0+func(index+1,1,prices,cap)
                dp[index][buy][cap]=max(buy_option,not_buy)
                return dp[index][buy][cap]
            else:
                sell=prices[index]+func(index+1,1,prices,cap-1)
                not_sell=0+func(index+1,0,prices,cap)
                dp[index][buy][cap]=max(sell,not_sell)
                return dp[index][buy][cap]
        return func(0,1,prices,k)


        
```
