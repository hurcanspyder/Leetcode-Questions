# Trick

*now here also we have used a concept of longest palindromic subsequence(LPS) because the elements who are already in a palindromic way not need to chnage but the elements who are not palindromic in the inital string need chnages*

**So answer is that the number of elements who are not palindromic are the ways to change a particular string into a palindrome**

```python
class Solution(object):
    def minInsertions(self, s):
        """
        :type s: str
        :rtype: int
        """
        a=s[::-1]
        def func(a,s):
            n=len(s)
            dp=[[0 for _ in range(n+1)]for _ in range(n+1)]

            for i in range(1,n+1):
                for j in range(1,n+1):
                    if s[i-1]==a[j-1]:
                        dp[i][j]=1+dp[i-1][j-1]
                    else:
                        dp[i][j]=0+max(dp[i-1][j],dp[i][j-1])
            return dp[-1][-1]
        return len(s)-func(a,s)
```
