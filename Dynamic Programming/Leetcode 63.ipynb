```python
class Solution(object):
    def uniquePathsWithObstacles(self, obstacleGrid):
        """
        :type obstacleGrid: List[List[int]]
        :rtype: int
        """
        def func(m,n,obstacleGrid,dp):
            if obstacleGrid[0][0]==1:
                return 0
            dp[0][0]=1
            for i in range(0,m):
                for j in range(0,n):
                    if i==0 and j==0:

                        continue
                    if obstacleGrid[i][j]==1:
                        dp[i][j]=0
                        continue
                    if i>0:
                        up=dp[i-1][j]
                    else:
                        up=0
                    if j>0:
                        left=dp[i][j-1]
                    else:
                        left=0
                    dp[i][j]=up+left
            return dp[m-1][n-1]
            

        dp=[[-1 for _ in range(len(obstacleGrid[0]))]for _ in range(len(obstacleGrid))]
        return func(len(obstacleGrid),len(obstacleGrid[0]),obstacleGrid,dp)
```
