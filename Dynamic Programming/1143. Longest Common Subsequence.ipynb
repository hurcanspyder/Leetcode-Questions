```python
class Solution(object):
    def longestCommonSubsequence(self, text1, text2):
        """
        :type text1: str
        :type text2: str
        :rtype: int
        """
        def func(index1,index2,text1,text2):
            if index1<0 or index2<0:
                return 0
            if text1[index1]==text2[index2]:
                return 1+func(index1-1,index2-1,text1,text2)
            return 0+max(func(index1-1,index2,text1,text2),func(index1,index2-1,text1,text2))
        return func(len(text1)-1,len(text2)-1,text1,text2)
```

O(2^n X 2^m)

SC->O(n+m)

# Memoization

```python
class Solution(object):
    def longestCommonSubsequence(self, text1, text2):
        """
        :type text1: str
        :type text2: str
        :rtype: int
        """
        def func(index1,index2,text1,text2,dp):
            if index1<0 or index2<0:
                return 0
            if dp[index1][index2]!=-1:
                return dp[index1][index2]
            if text1[index1]==text2[index2]:
                dp[index1][index2]=1+func(index1-1,index2-1,text1,text2,dp)
                return dp[index1][index2]
            dp[index1][index2]= 0+max(func(index1-1,index2,text1,text2,dp),func(index1,index2-1,text1,text2,dp))
            return dp[index1][index2]
        dp=[[-1 for _ in range(len(text2))]for _ in range(len(text1))]
        return func(len(text1)-1,len(text2)-1,text1,text2,dp)
```

TC->O(n X m)

SC->O(n+m)+O(n X m)

```python
class Solution(object):
    def longestCommonSubsequence(self, text1, text2):
        """
        :type text1: str
        :type text2: str
        :rtype: int
        """
        def tabulation(text1,text2,dp):
            for ind1 in range(0,len(text1)+1):
                dp[ind1][0]=0
            for ind2 in range(0,len(text2)+1):
                dp[0][ind2]=0
            for index1 in range(1,len(text1)+1):
                for index2 in range(1,len(text2)+1):
                    if text1[index1-1]==text2[index2-1]:
                        dp[index1][index2]=1+dp[index1-1][index2-1]
                    else:
                        dp[index1][index2]=0+max(dp[index1-1][index2],dp[index1][index2-1])
            return dp[-1][-1]

        dp=[[-1 for _ in range(len(text2)+1)]for _ in range(len(text1)+1)]
        return tabulation(text1,text2,dp)
```

# Tabularion with space optimization

```python
class Solution:
    def longestCommonSubsequence(self, text1: str, text2: str) -> int:
        n = len(text1)
        m = len(text2)
        prev = [-1 for _ in range(m + 1)]
        for ind2 in range(0, m + 1):
            prev[ind2] = 0
        # for ind1 in range(0,n+1):
        #     dp[ind1][0]=0
        for ind1 in range(1, n + 1):
            curr = [-1 for _ in range(m + 1)]
            curr[0] = 0
            for ind2 in range(1, m + 1):
                if text1[ind1 - 1] == text2[ind2 - 1]:
                    curr[ind2] = 1 + prev[ind2 - 1]
                else:
                    curr[ind2] = 0 + max(prev[ind2], curr[ind2 - 1])
            prev = curr
        return prev[m]
```
