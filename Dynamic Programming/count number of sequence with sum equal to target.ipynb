# Memoization

```python
def func(index,nums,target,dp):
    if target==0:
        return True
    if index==0:
        if nums[0]==target:
            return True
        return False
    if dp[index][target]!=-1:
        return dp[index][target]
    
    if nums[index]>target:
        pick=False
    else:
        pick=func(index-1,nums,target-nums[index],dp)
    not_pick=func(index-1,nums,target,dp)
    dp[index][target]=pick or not_pick
    return dp[index][target]
nums=[1,4,3,45,46,7]
target=7

dp=[[-1 for _ in range(target+1)]for _ in range(len(nums))]
func(5,nums,target,dp)
count=0
for i in range(0,len(nums)):
    if dp[i][target]==True:
        count+=1
print(count)
print(dp)
```
