```python
class Solution(object):
    def maximalSquare(self, matrix):
        """
        :type matrix: List[List[str]]
        :rtype: int
        """
        rows=len(matrix)
        cols=len(matrix[0])
        maxi=0
        flag=False
        for i in range(0,rows):
            for j in range(0,cols):
                if matrix[i][j]=="1":
                    flag=True
                    break
        
        for i in range(1,rows):
            for j in range(1,cols):
               
                if matrix[i][j]=="1":
                    print("h")
                    matrix[i][j]=1+min(int(matrix[i-1][j-1]),int(matrix[i-1][j]),int(matrix[i][j-1]))
                    maxi=max(matrix[i][j],maxi)
        if maxi==0:
            if flag:
                return 1
            else:
                return 0
        else:
            return maxi*maxi

```
