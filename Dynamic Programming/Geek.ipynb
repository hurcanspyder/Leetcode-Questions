# Recursion

```python
def func(day,last,arr):
    if day==0:
        maxi=0
        for i in range(0,3):
            if i!=last:
                maxi=max(maxi,arr[day][i])
        return maxi
    maxi=0
    for i in range(0,3):
        if i!=last:
            maxi=max(maxi,arr[day][i]+func(day-1,i,arr))
    return maxi
```

# Memoization

```python
def func(day,last,arr,dp):
    if day==0:
        maxi=0
        for i in range(0,3):
            if i!=last:
                maxi=max(maxi,arr[day][i])
        return maxi
    if dp!=-1:
        return[day][last]
    maxi=0

    for i in range(0,3):
        if i!=last:
            maxi=max(maxi,arr[day][i]+func(day-1,i,arr,dp))
    dp[day][last]=maxi
    return dp[day][last]
```
