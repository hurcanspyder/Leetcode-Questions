Its a DFS problem

```python
class Solution(object):
    def lexicalOrder(self, n):
        """
        :type n: int
        :rtype: List[int]
        """
        if n<=9:
            return [i for i in range(1,n+1)]
        list1=[]
        curr=1
        for _ in range(n):
            list1.append(curr)
            if curr*10<=n:
                curr*=10
            
            else:
                if curr%10==9 or curr+1>n:
                    curr//=10
                curr+=1
        return list1
        
      
```
