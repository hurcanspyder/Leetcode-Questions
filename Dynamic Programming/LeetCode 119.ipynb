```python
class Solution(object):
    def getRow(self, rowIndex):
        """
        :type rowIndex: int
        :rtype: List[int]
        """

        def func(rowIndex):
            if rowIndex==0:
                return [1]
            prevrow=func(rowIndex-1)
            newrow=[1]*(rowIndex+1)
            for i in range(1,rowIndex):
                newrow[i]=prevrow[i-1]+prevrow[i]
            prevrow=newrow
            return prevrow
        return func(rowIndex)
```
