```python
class Solution(object):
    def countNumbersWithUniqueDigits(self, n):
        """
        :type n: int
        :rtype: int
        """
        if n==0:
            return 1
        def func(n):
            if n==1:
                return 10
            
            
            count=9
            prod=9
            
            while n-1>0:

                count*=(prod)
                n-=1
                prod-=1
            return count
        a=0
        while n>0:
            a+=func(n)
            n-=1
        return a
```

# DP version

```python
class Solution(object):
    def countNumbersWithUniqueDigits(self, n):
        if n == 0:
            return 1

        # dp[i] = count of unique-digit numbers with length i
        dp = [0] * (n + 1)
        dp[1] = 10

        available = 9   # for second digit (after choosing first digit)
        
        for i in range(2, n + 1):
            dp[i] = dp[i - 1] * available
            available -= 1

        return sum(dp[1:])
```
