# Memory llimit get exceeded if n>=2147483647

```python
class Solution(object):
    def integerReplacement(self, n):
        if n == 1:
            return 0
        
        dp = [0] * (n + 2)    # +2 for safety
        
        dp[1] = 0
        
        for i in range(2, n + 1):
            if i % 2 == 0:
                dp[i] = 1 + dp[i // 2]
            else:
                dp[i] = 1 + min(dp[i - 1], dp[(i + 1) // 2] + 1)  # simulate +1 then /2
        
        return dp[n]


        
```

# optimal way

```python
class Solution(object):
    def integerReplacement(self, n):
        memo = {1: 0}
        
        def solve(x):
            if x in memo:
                return memo[x]

            if x % 2 == 0:
                memo[x] = 1 + solve(x // 2)
            else:
                memo[x] = 1 + min(solve(x + 1), solve(x - 1))
            
            return memo[x]

        return solve(n)
```
