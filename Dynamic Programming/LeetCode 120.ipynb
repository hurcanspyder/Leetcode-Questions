*In this question one end is variable and another end is fixed means starting point is fixed and end point is variable and if we want to solve this question easily we have to start the reccursion with a fixed side no matter where this side lie in our case fixed side is starting point means we will start our recurssion process from the star of the triangle*

```python
def func(i,j,triangle):
    if i==0 and j==0:
        return triangle[i][j]
    
    
    diag=triangle[i][j]+func(i-1,j-1,triangle)
    left=triangle[i][j]+func(i,j-1,triangle)
    return min(diag,left)







    


```

```python
class Solution(object):
    def minimumTotal(self, triangle):
        """
        :type triangle: List[List[int]]
        :rtype: int
        """
        def func(i,j,triangle,dp):
            if i==len(triangle)-1:
                return triangle[i][j]
            if dp[i][j]!=-1:
                return dp[i][j]
            down=triangle[i][j]+func(i+1,j,triangle,dp)
            diag=triangle[i][j]+func(i+1,j+1,triangle,dp)
            dp[i][j]= min(down,diag)
            return dp[i][j]
        dp=[[-1 for _ in range(len(triangle[-1]))]for _ in range(len(triangle))]
        return func(0,0,triangle,dp)
```

# CHat gpt solution

```python
class Solution:
    def minimumTotal(self, triangle):
        n = len(triangle)
        dp = triangle[-1][:]  # copy last row

        for i in range(n - 2, -1, -1):
            for j in range(len(triangle[i])):
                dp[j] = triangle[i][j] + min(dp[j], dp[j+1])

        return dp[0]
```
