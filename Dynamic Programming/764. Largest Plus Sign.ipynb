```python
class Solution(object):
    def orderOfLargestPlusSign(self, n, mines):

        grid = [[1 for _ in range(n)] for _ in range(n)]
        for x, y in mines:
            grid[x][y] = 0

        dp = [[[0, 0] for _ in range(n)] for _ in range(n)]

        # Up + Left
        for i in range(n):
            for j in range(n):
                if grid[i][j] == 1:
                    up = dp[i-1][j][0] if i > 0 else 0
                    left = dp[i][j-1][1] if j > 0 else 0
                    dp[i][j] = [up + 1, left + 1]

        dp1 = [[[0, 0] for _ in range(n)] for _ in range(n)]

        # Down + Right
        for i in range(n-1, -1, -1):
            for j in range(n-1, -1, -1):
                if grid[i][j] == 1:
                    down = dp1[i+1][j][0] if i < n-1 else 0
                    right = dp1[i][j+1][1] if j < n-1 else 0
                    dp1[i][j] = [down + 1, right + 1]

        maxi = 0

        # Combine all 4 directions
        for i in range(n):
            for j in range(n):
                order = min(dp[i][j][0], dp[i][j][1],
                            dp1[i][j][0], dp1[i][j][1])
                maxi = max(maxi, order)

        return maxi


        

        

        

        
        

        
        


        
```
