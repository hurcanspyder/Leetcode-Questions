# Recursion

```python
class Solution(object):
    def maxProfit(self, prices, fee):
        """
        :type prices: List[int]
        :type fee: int
        :rtype: int
        """
        def func(index,x,prices,fee):
            if index>=len(prices):
                return 0
            if x==0:
                buy = - prices[index]+func(index+1,1,prices,fee)
                not_buy=0+func(index+1,0,prices,fee)
                last_buy=max(buy,not_buy)
                
            if x==1:
                sell=+prices[index]+func(index+1,0,prices,fee)-fee
                not_sell=0+func(index+1,1,prices,fee)
                last_buy=max(sell,not_sell)
                
            return last_buy
        return func(0,0,prices,fee)

            
            

            



        
```

# DP

```python
class Solution(object):
    def maxProfit(self, prices, fee):
        """
        :type prices: List[int]
        :type fee: int
        :rtype: int
        """
        def func(index, x, prices, fee, dp):
            if index >= len(prices):
                return 0

            if dp[index][x] != -1:
                return dp[index][x]

            if x == 0:
                buy = -prices[index] + func(index+1, 1, prices, fee, dp)
                not_buy = func(index+1, 0, prices, fee, dp)
                dp[index][x] = max(buy, not_buy)

            else:
                sell = prices[index] + func(index+1, 0, prices, fee, dp) - fee
                not_sell = func(index+1, 1, prices, fee, dp)
                dp[index][x] = max(sell, not_sell)

            return dp[index][x]
        dp=[[-1 for _ in range(2)]for _ in range(len(prices)+1)]
        return func(0,0,prices,fee,dp)
        


            
            

            



        
```
