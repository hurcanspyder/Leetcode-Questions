```python
class Solution(object):
    def coinChange(self, coins, amount):
        if amount==0:
            return 0
        def combinationsum(subset, total, index, nums, target, result):

            if total == target:
                result.append(subset[:])
                return
            if index >= len(nums) or total > target:
                return
            # choose coin
            subset.append(nums[index])
            combinationsum(subset, total + nums[index], index, nums, target, result)
            subset.pop()
            # skip coin
            combinationsum(subset, total, index + 1, nums, target, result)
            return result

        a = combinationsum([], 0, 0, coins, amount, [])
        if not a:
            return -1
        return min(len(x) for x in a)   # minimum number of coins


        
        
```

# Recursion

```python
class Solution(object):
    def coinChange(self, coins, amount):
        if amount==0:
            return 0
        def func(index,amount,coins):
            if amount==0:
                return 0
            if index==0:
                if amount%coins[0]==0:
                    return amount//coins[0]
                return 10**8
            if coins[index]>amount:
                pick=10**8
            else:
                pick=1+func(index,amount-coins[index],coins)
            not_pick=func(index-1,amount,coins)
            return min(pick,not_pick)
        a= func(len(coins)-1,amount,coins)
        if a!=10**8:
            return a
        return -1
```

# DP

```python
class Solution(object):
    def coinChange(self, coins, amount):
        if amount==0:
            return 0
        def func(index,amount,coins,dp):
            if amount==0:
                return 0
            if index==0:
                if amount%coins[0]==0:
                    return amount//coins[0]
                return 10**8
            if dp[index][amount]!=-1:
                return dp[index][amount]
            if coins[index]>amount:
                pick=10**8
            else:
                pick=1+func(index,amount-coins[index],coins,dp)
            not_pick=func(index-1,amount,coins,dp)
            dp[index][amount]= min(pick,not_pick)
            return dp[index][amount]
        dp=[[-1 for _ in range(amount+1)]for _ in range(len(coins))]
        a= func(len(coins)-1,amount,coins,dp)
        if a!=10**8:
            return a
        return -1


            


        
        
```
