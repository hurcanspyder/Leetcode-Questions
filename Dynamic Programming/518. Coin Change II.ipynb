# Recursion

```python
class Solution(object):
    def change(self, amount, coins):
        """
        :type amount: int
        :type coins: List[int]
        :rtype: int
        """
        def func(index,amount,coins):
            if amount==0:
                return 1
            if index==0:
                if amount%coins[0]==0:
                    return 1
                return 0
            if coins[index]>amount:
                pick=0
            else:
                pick=func(index,amount-coins[index],coins)
            not_pick=func(index-1,amount,coins)
            return pick+not_pick
        return func(len(coins)-1,amount,coins)
        
```

# DP

```python
class Solution(object):
    def change(self, amount, coins):
        """
        :type amount: int
        :type coins: List[int]
        :rtype: int
        """
        def func(index,amount,coins,dp):
            if amount==0:
                return 1
            if index==0:
                if amount%coins[0]==0:
                    return 1
                return 0
            if dp[index][amount]!=-1:
                return dp[index][amount]
            if coins[index]>amount:
                pick=0
            else:
                pick=func(index,amount-coins[index],coins,dp)
            not_pick=func(index-1,amount,coins,dp)
            dp[index][amount]=pick+not_pick
            return dp[index][amount]
        dp=[[-1 for _ in range(amount+1)]for _ in range(len(coins))]
        return func(len(coins)-1,amount,coins,dp)
```

# Lovely Tabulation

```python
class Solution(object):
    def change(self, amount, coins):
        n = len(coins)
        dp = [[0 for _ in range(amount + 1)] for _ in range(n)]

        # Base case: amount = 0
        for i in range(n):
            dp[i][0] = 1

        # Base case: using only first coin
        for a in range(amount + 1):
            if a % coins[0] == 0:
                dp[0][a] = 1

        # Fill the table
        for i in range(1, n):
            for a in range(1, amount + 1):
                not_pick = dp[i - 1][a]
                pick = 0
                if coins[i] <= a:
                    pick = dp[i][a - coins[i]]
                dp[i][a] = pick + not_pick

        return dp[n - 1][amount]
```
