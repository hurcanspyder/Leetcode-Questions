```python
class Solution(object):
    def maximizeSquareHoleArea(self, n, m, hBars, vBars):
        """
        :type n: int
        :type m: int
        :type hBars: List[int]
        :type vBars: List[int]
        :rtype: int
        """
        hBars.sort()
        vBars.sort()
        maxi=1
        count=1
        count1=1
        maxi1=1
        
        for i in range(1,len(hBars)):
            if hBars[i]==hBars[i-1]+1:
                count+=1
            else:
                count=1
            maxi=max(count,maxi)
        for j in range(1,len(vBars)):
            if vBars[j]==vBars[j-1]+1:
                count1+=1
            else:
                count1=1
            maxi1=max(count1,maxi1)
        print(maxi,maxi1)
        mini=min(maxi,maxi1)+1
        return mini*mini
        
```
