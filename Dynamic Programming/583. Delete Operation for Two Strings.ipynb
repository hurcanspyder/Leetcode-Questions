```python
class Solution(object):
    def minDistance(self, word1, word2):
        """
        :type word1: str
        :type word2: str
        :rtype: int
        """
        def func(word1,word2):
            m=len(word1)
            n=len(word2)
            dp=[[0 for _ in range(n+1)]for _ in range(m+1)]
            for i in range(1,m+1):
                for j in range(1,n+1):
                    if word1[i-1]==word2[j-1]:
                        dp[i][j]=1+dp[i-1][j-1]
                    else:
                        dp[i][j]=0+max(dp[i-1][j],dp[i][j-1])
            return dp[-1][-1]
        a=func(word1,word2)
        count=0
        if a<len(word1):
            count=count+(len(word1)-a)
        if a<len(word2):
            count=count+(len(word2)-a)
        return count

        
```
