# By recurssion

```python
class Solution(object):
    def combinationSum4(self, nums, target):

        count = [0]

        def backtrack(current_sum, subset):
            if current_sum == target:
                count[0] += 1
                return
            
            if current_sum > target:
                return
            
            for x in nums:
                subset.append(x)
                backtrack(current_sum + x, subset)
                subset.pop()     # undo choice

        backtrack(0, [])
        return count[0]



        
```

```python
def func(nums,target,dp):  
    dp=[1]  
    for i in range(1, target + 1):
        total = 0
        for x in nums:
            if i - x >= 0:
                total += dp[i - x]
        dp.append(total)
    return dp[-1]
```

| Part    | Meaning                   |
| ------- | ------------------------- |
| dp      | stores ways to reach sums |
| dp[0]=1 | one way to make sum 0     |
| dp[i]   | sum of previous dp values |
| dp[-1]  | result for target         |
