```python
class Solution(object):
    def cherryPickup(self, grid):
        """
        :type grid: List[List[int]]
        :rtype: int
        """
        def func(i,j1,j2,grid):
            if j2<0 or j1<0  or j1>len(grid[0])-1 or j2>len(grid[0])-1 :
                return -10**7
            if i==len(grid)-1:
                if j1==j2:
                    return grid[i][j1]
                
                return grid[i][j1]+grid[i][j2]
            maxi=0
            for new_j1 in range(-1,2):
                for new_j2 in range(-1,2):
                    if j1==j2:
                        ans=grid[i][j1]+func(i+1,j1+new_j1,j2+new_j2,grid)
                    else:
                        ans=grid[i][j1]+grid[i][j2]+func(i+1,j1+new_j1,j2+new_j2,grid)
                    maxi=max(maxi,ans)
            return maxi

        return func(0,0,len(grid[0])-1,grid)
```
