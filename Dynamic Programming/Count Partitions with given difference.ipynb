# Reccursion

```python
def func(index,nums,target):
    if target==0:
        return True
    if index==0:
        if nums[0]==target:
            return True
        return False
    
    if nums[index]>target:
        pick=False
    else:
        pick=func(index-1,nums,target-nums[index])
    not_pick=func(index-1,nums,target)
    return pick or not_pick
   
    
```

```python
def solve():
    target=(sum([1,2,4,6,2,55])-1000)/2
    if target<0 or target%2!=0:
        return False
    return func(5,[1,2,4,6,2,55],target)
solve()
```

# Memoization

```python
def func1(index,nums,target,dp):
    if target==0:
        return True
    if index==0:
        if nums[0]==target:
            return True
        return False
    if dp[index][target]!=-1:
        return dp[index][target]
    if nums[index]>target:
        pick=False
    else:
        pick=func1(index-1,nums,target-nums[index],dp)
    not_pick=func1(index-1,nums,target,dp)
    dp[index][target]=pick or not_pick
    return dp[index][target]
```

```python
def solve():
    nums=[1,2,4,6,2,55]
    d=int(input("Enter the difference you want to check"))
    n=len(nums)
    target=(sum(nums)-d)/2
    if target<0 or (sum(nums)-d)%2!=0:
        print("h")
        return False
    dp=[[-1 for _ in range(int(target)+1)]for _ in range(n)]
    return func1(n-1,nums,int(target),dp)
solve()
```
