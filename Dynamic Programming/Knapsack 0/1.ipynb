```python
def func(index,w,val,weight):
    if index==0:
        if weight[index]<=w:
            return weight[index]
        return 0

    if weight[index]>w:
        pick=-10**7
    else:

        pick=val[index]+func(index-1,w-weight[index],val,weight)
    not_pick=0+func(index-1,w,val,weight)
    return max(pick,not_pick)

```

```python
def knapsack(index,w,val,wt):
    n=len(wt)
    dp=[[0 for _ in range(0,w+1)]for _ in range(n)]
    for w_ in range(0,w):
        if wt[0]<=w:
            dp[0][w_]=val[0]
        for index in range(1,n):
            for w_ in range(0,w+1):
                if wt[index]>w_:
                    pick=-10**7
                else:
                    pick=val[index]+dp[index-1][w_-w[index]]
                    not_pick=0+dp[index-1][w_]
                dp[index][w_]=max(pick,not_pick)
        return dp[n-1][w]
        
```
