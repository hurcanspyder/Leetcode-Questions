```python
class Solution(object):
    def findMaxForm(self, strs, m, n):
        """
        :type strs: List[str]
        :type m: int
        :type n: int
        :rtype: int
        """
        list1=[]
        for i in range(0,len(strs)):
            list2=[0,0]
            for j in strs[i]:
                list2[int(j)]+=1
            list1.append(list2)
        dp=[[0 for _ in range(n+1)]for _ in range(m+1)]
        stack=[]
        for x,y in list1:
            for s1 in range(m,x-1,-1):
                for s2 in range(n,y-1,-1):
                    # if s1>=x and s2>=y:
                    dp[s1][s2]=max(dp[s1][s2],dp[s1-x][s2-y]+1)
        return dp[m][n]
```
