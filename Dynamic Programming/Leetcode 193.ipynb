# Reccursion

```python
def func(index,nums):
    if index==0:
        return nums[index]
    if index<0:
        return 0
    pick=nums[index]+func(index-2,nums)
    not_pick=0+func(index-1,nums)
    return max(pick,not_pick)
```

# Memoization

```python
def func(index,nums,dp):
    if index==0:
        return nums[index]
    if index<0:
        return 0
    if dp[index]!=-1:
        return dp[index]
    pick=nums[index]+func(index-2,nums,dp)
    not_pick=0+func(index-1,nums,dp)
    dp[index]= max(pick,not_pick)
```

*The difference between recurssion and tabulation is that in recursion we are calling function againa and again but in tabulation we are calling dp array agin and again (Here we will start from the bottom and go up to top to achieve the desired result*-----> instead of doing func(index-2) ---------> dp[index-2]

# Tabulation

```python
def func3(dp,nums):
    dp[0]=nums[0]
    for index in range(1,len(nums)):
        if index>1:
            pick=nums[index]+dp[index-2]
        else:
            pick=nums[index]
        not_pick=0+dp[index-1]
        dp[index]=max(pick,not_pick)
    return dp[-1]
```

# Tabulation (space optimization)

```python
def func4(nums):
    prev=0
    prev1=nums[0]
    curr=0
    for index in range(1,len(nums)):
        if index>1:
            pick=nums[index]+prev
        else:
            pick=nums[index]
        not_pick=prev1
        curr=max(pick,not_pick)
        prev=prev1
        prev1=curr
    return prev1
        
    
```
