```python
class Solution(object):
    def climbStairs(self, n):
        """
        :type n: int
        :rtype: int
        """
        def func(n):
            if n==0:
                return 1
            if n==1:
                return 1
            prev=1
            prev1=1
            curr=0
            for i in range(2,n+1):
                curr=prev+prev1
                prev1=prev
                prev=curr
            return prev

                

      
      
        return func(n)


        
```
