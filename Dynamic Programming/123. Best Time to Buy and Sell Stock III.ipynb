```python
class Solution(object):
    def maxProfit(self, prices):
        n = len(prices)
        dp = [[[-1 for _ in range(3)] for _ in range(2)] for _ in range(n)]

        def func(i, buy, cap):
            if i == n or cap == 0:
                return 0

            if dp[i][buy][cap] != -1:
                return dp[i][buy][cap]

            if buy == 1:
                take = -prices[i] + func(i+1, 0, cap)
                skip = func(i+1, 1, cap)
                dp[i][buy][cap] = max(take, skip)
            else:
                sell = prices[i] + func(i+1, 1, cap-1)
                skip = func(i+1, 0, cap)
                dp[i][buy][cap] = max(sell, skip)

            return dp[i][buy][cap]

        return func(0, 1, 2)

```
