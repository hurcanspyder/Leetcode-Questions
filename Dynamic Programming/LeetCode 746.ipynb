```python
class Solution(object):
    def minCostClimbingStairs(self, cost):
        """
        :type cost: List[int]
        :rtype: int
        """
        def func(n,cost,dp):
            if n==0 or n==1:
                return cost[n]
            if n<0:
                return 0
            if dp[n]!=-1:
                return dp[n]
            step1=cost[n]+func(n-1,cost,dp)
            step2=cost[n]+func(n-2,cost,dp)
            dp[n]= min(step1,step2)
            return dp[n]
        n=len(cost)
        dp=[-1 for _ in range(len(cost))]
        return min(func(n-1,cost,dp),func(n-2,cost,dp))
        
```
