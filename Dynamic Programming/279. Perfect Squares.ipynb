# Goood code

```python
class Solution(object):
    def numSquares(self, n):
        list1 = []
        for i in range(1, int(n**0.5)+1):
            list1.append(i*i)

        def func(index, nums, target, dp):
            if target == 0:
                return 0
            if index == 0:
                if target % nums[0] == 0:
                    return target // nums[0]
                else:
                    return 10**8

            if dp[index][target] != -1:
                return dp[index][target]

            # pick
            pick = 10**8
            if nums[index] <= target:
                pick = 1 + func(index, nums, target - nums[index], dp)

            # not pick
            not_pick = func(index - 1, nums, target, dp)

            dp[index][target] = min(pick, not_pick)
            return dp[index][target]

        dp = [[-1 for _ in range(n + 1)] for _ in range(len(list1))]
        return func(len(list1)-1, list1, n, dp)
```

# The Best one 

```python
class Solution(object):
    def numSquares(self, n):

        dp = [float('inf')] * (n+1)
        dp[0] = 0

        for i in range(1, n+1):
            j = 1
            while j*j <= i:
                dp[i] = min(dp[i], 1 + dp[i - j*j])
                j += 1

        return dp[n]
```
