# Tabulation

```python
class Solution(object):
    def minPathSum(self, grid):
        """
        :type grid: List[List[int]]
        :rtype: int
        """
        def func(m,n,grid,dp):
            dp[0][0]=grid[0][0]
            for i in range(0,m):
                for j in range(0,n):
                    if i==0 and j==0:
                        continue
                    if i>0:
                        up=grid[i][j]+dp[i-1][j]
                    else:
                        up=10**8
                    if j>0:
                        left=grid[i][j]+dp[i][j-1]
                    else:
                        left=10**8
                    dp[i][j]=min(up,left)
            return dp[m-1][n-1]
        m=len(grid)
        n=len(grid[0])
        dp=[[-1 for _ in range(n)]for _ in range(m)]
        return func(m,n,grid,dp)        
```

# Tabulation with space optimization


```python
class Solution(object):
    def minPathSum(self, grid):
        """
        :type grid: List[List[int]]
        :rtype: int
        """
        def func(m,n,grid):
            prev=[0]*n
            for i in range(0,m):
                curr=[-1]*n
                for j in range(0,n):
                    if i==0 and j==0:
                        continue
                    if i>0:
                        up=grid[i][j]+prev[j]
                    else:
                        up=10**8
                    if j>0:
                        left=grid[i][j]+curr[j-1]
                    else:
                        left=10**8
                    curr[j]=min(up,left)
                prev=curr.copy()
            return prev[n-1]
        m=len(grid)
        n=len(grid[0])
        dp=[[-1 for _ in range(n)]for _ in range(m)]
        return func(m,n,grid)        
```

Here dp[i-1]=prev and dp[i]=curr
