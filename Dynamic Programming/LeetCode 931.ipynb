# Memoization



```python
class Solution(object):
    def minFallingPathSum(self, matrix):
        """
        :type matrix: List[List[int]]
        :rtype: int
        """
        def func(i,j,matrix,dp):
            if j<0 or j>len(matrix[0])-1:
                return 10**8
            if i==0:
                return matrix[i][j]
            if dp[i][j]!=-1:
                return dp[i][j]
            left=matrix[i][j]+func(i-1,j-1,matrix,dp)
            up=matrix[i][j]+func(i-1,j,matrix,dp)
            right=matrix[i][j]+func(i-1,j+1,matrix,dp)
            dp[i][j]= min(left,up,right)
            return dp[i][j]
        n=len(matrix[0])
        m=len(matrix)
        mini=10**8
        dp=[[-1 for _ in range(n)]for _ in range(m)]

        for i in range(n):
            
            mini=min(mini,func(m-1,i,matrix,dp))
        return mini
```

```python
class Solution(object):
    def minFallingPathSum(self, matrix):
        """
        :type matrix: List[List[int]]
        :rtype: int
        """
        def func(m,n,matrix,dp):
            dp[0]=matrix[0][:]
            for i in range(1,m):
                for j in range(0,n):
                    if j>0:
                        left=matrix[i][j]+dp[i-1][j-1]
                    else:
                        left=10**8
                    if j<n-1:
                        right=matrix[i][j]+dp[i-1][j+1]
                    else:
                        right=10**8
                    down=matrix[i][j]+dp[i-1][j]
                    dp[i][j]=min(left,right,down)
            return min(dp[-1])


        n=len(matrix[0])
        m=len(matrix)
        mini=10**8
        dp=[[-1 for _ in range(n)]for _ in range(m)]
        return func(m,n,matrix,dp) 
```
