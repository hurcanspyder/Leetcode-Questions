```python
class Solution(object):
    def findPaths(self, m, n, maxMove, startRow, startColumn):
        """
        :type m: int
        :type n: int
        :type maxMove: int
        :type startRow: int
        :type startColumn: int
        :rtype: int
        """
        MOD=10**9+7
        def func(i,j,maxMove,count,m,n,dp):
            if i<0 or i>=m or j<0 or j>=n:
                return 1
            if count>=maxMove:
                return 0
            if dp[i][j][count]!=-1:
                return dp[i][j][count]
            left=func(i,j-1,maxMove,count+1,m,n,dp)
            right=func(i,j+1,maxMove,count+1,m,n,dp)
            up=func(i-1,j,maxMove,count+1,m,n,dp)
            down=func(i+1,j,maxMove,count+1,m,n,dp)
            dp[i][j][count]= (left+right+up+down)%MOD
            return dp[i][j][count]
        dp=[[[-1 for _ in range(maxMove+1)]for _ in range(n)]for _ in range(m)]
        return func(startRow,startColumn,maxMove,0,m,n,dp)
        
```
