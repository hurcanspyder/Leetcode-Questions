# Same as house robber question whenever there is a question of neighbour hood do remember these two questions

```python
class Solution:
    def deleteAndEarn(self, nums):
        if  not nums:
            return 0

        freq = [0] * (max(nums)+1)
        for n in nums:
            freq[n] += n

        dp = [0] * len(freq)
        dp[1] = freq[1]
        for i in range(2, len(freq)):
            dp[i] = max(freq[i] + dp[i-2], dp[i-1])
            

        return dp[len(freq)-1]
        
        
```
