```python
def tabulation(text1,text2 ):
    n=len(text1)
    m=len(text2)

    dp=[[0 for _ in range(m+1)]for _ in range(n+1)]
    for i in range(1,n+1):
        for j in range(1,n+1):
            if text1[i-1]==text2[j-1]:
                dp[i][j]=1+dp[i-1][j-1]
            else:
                dp[i][j]=max(dp[i-1][j],dp[i][j-1])
    print(dp)
    i,j=n,m
    result=[]
    while i>0 and j>0:
        if text1[i-1]==text2[j-1]:
            result.append(text1[i-1])
            i-=1
            j-=1

        else:
            if dp[i][j]>=dp[i-1][j]:
                j-=1
            else:
                i-=1
    result.reverse()
    print("".join(result))


```
