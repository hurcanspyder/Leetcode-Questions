```python
class Solution(object):
    def rob(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        if len(nums)==1:
            return nums[0]
        def func(index,nums,dp):
            if index<0:
                return 0
            if dp[index]!=-1:
                return dp[index]
            pick=nums[index]+func(index-2,nums,dp)
            not_pick=0+func(index-1,nums,dp)

            dp[index]= max(pick,not_pick)
            return dp[index]
        new_nums=nums[1:]
        dp1=[-1 for _ in range(len(new_nums))]
        dp=[-1 for _ in range(len(nums))]
        return max(func(len(new_nums)-1,new_nums,dp1),func(len(nums)-2,nums,dp))
            

        
```
