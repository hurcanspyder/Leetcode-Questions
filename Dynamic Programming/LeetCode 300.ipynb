# Memoization

```python
class Solution(object):
    def lengthOfLIS(self, nums):
        n = len(nums)
        
        from functools import lru_cache
        
        @lru_cache(None)
        def solve(index, prev_index):
            if index == n:
                return 0
            
            # option 1: skip current number
            best = solve(index + 1, prev_index)
            
            # option 2: include nums[index] only if increasing
            if prev_index == -1 or nums[index] > nums[prev_index]:
                best = max(best, 1 + solve(index + 1, index))
            
            return best
        
        return solve(0, -1)
```

# Tabulation

```python
class Solution(object):
    def lengthOfLIS(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        n=len(nums)
        dp=[1]*n
        max_len=1
        for i in range(n):
            for j in range(i):
                if nums[i]>nums[j]:
                    dp[i]=max(dp[i],dp[j]+1)
                max_len=max(max_len,dp[i])
        return max_len

        
```
