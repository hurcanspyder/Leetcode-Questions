```python
class Solution(object):
    def maxProfit(self, prices):
        def func(index, prices, buy,dp):
            
            if index == len(prices):
                return 0
            if dp[index][buy]!=-1:
                return dp[index][buy]

            if buy == 1:
                buy_option = -prices[index] + func(index + 1, prices, 0,dp)
                skip_option = func(index + 1, prices, 1,dp)
                dp[index][buy]= max(buy_option, skip_option)
                return dp[index][buy]
            else:
                sell_option = prices[index] + func(index + 1, prices, 1,dp)
                skip_option = func(index + 1, prices, 0,dp)
                dp[index][buy]= max(sell_option, skip_option)
                return dp[index][buy]
        dp=[[-1 for _ in range(2)]for _ in range(len(prices))]
        return func(0, prices, 1,dp)

        
```
