```python
class Solution(object):
    def minimumDeleteSum(self, s1, s2):
        """
        :type s1: str
        :type s2: str
        :rtype: int
        """
        ord1=0
        for i in s1:
            ord1+=ord(i)
        for j in s2:
            ord1+=ord(j)
        def func(index1,index2,s1,s2,dp):
            if index1<0 or index2<0:
                return 0
            if dp[index1][index2]!=-1:
                return dp[index1][index2]
            if s1[index1]==s2[index2]:
                dp[index1][index2]=ord(s1[index1])+func(index1-1,index2-1,s1,s2,dp)
                return dp[index1][index2]

            else:
                dp[index1][index2]=max(func(index1-1,index2,s1,s2,dp),func(index1,index2-1,s1,s2,dp))
                return dp[index1][index2]
            
        dp=[[-1 for _ in range(len(s2))]for _ in range(len(s1))]
        return ord1-2*func(len(s1)-1,len(s2)-1,s1,s2,dp)

        
```
