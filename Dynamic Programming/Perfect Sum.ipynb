*Now the difference between perfect sum and normal sum is that in normal sum we are not couting 0 but in case perfect sum we gonna count 0 and negative(-ve) integers too*

# Recursive approach

```python
def func(index,total,nums):
    if index==0:
        if total==0 and nums[0]==0:
            return 2
        if total==0 or nums[0]==total:
            return 1
        return 0
    if nums[index]>total:
        return 0
    else:
        pick=func(index-1,total-nums[index],nums)
    not_pick=func(index-1,total,nums)
    return pick+not_pick
```

# Memoization approach

```python
def func(index,total,nums,dp):
    if index==0:
        if total==0 and nums[0]==0:
            return 2
        if total==0 or nums[0]==total:
            return 1
        return 0
    if dp[index][total]!=-1:
        return dp[index][total]
    if nums[index]>total:
        return 0
    else:
        pick=func(index-1,total-nums[index],nums,dp)
    not_pick=func(index-1,total,nums,dp)
    dp[index][total]=pick+not_pick
    return dp[index][total]
target=6
nums=[5,2,3,10,6,8]
dp=[[-1 for _ in range(target+1)]for _ in range(len(nums))]
func(5,target,nums,dp)

```
