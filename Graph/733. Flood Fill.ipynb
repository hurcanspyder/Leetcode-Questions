# DFS Approach

```python
def floodfill(image,sr,sc,color):
    if image[sr][sc]==color:
        return image
    import copy
    vis=copy.deepcopy(image)
    rows=len(vis)
    cols=len(vis[0])
    initial_color=vis[sr][sc]
    def dfs(i,j,new_color,initial_color,vis,rows,cols):
        if i<0 or i>=rows or j<0 or j>=cols:
            return
        if vis[i][j]!=initial_color:
            return
        if vis[i][j]==new_color:
            return
        vis[i][j]=new_color

        dfs(i+1,j)  #right movement
        dfs(i,j-1)  #upward movement
        dfs(i-1,j)  #left movement
        dfs(i,j+1)  #downward movement
    dfs(sr,sc,color,initial_color,vis,rows,cols)
    return vis
```

TC->O(rowsXcolsX4)

SC->O(RowsXcols)+O(rowsXcols)  one is stack space
