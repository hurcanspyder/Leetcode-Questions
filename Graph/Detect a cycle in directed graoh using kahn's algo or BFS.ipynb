**Its a hack here look we get all element in our result when we use Kahn's algo for topological sort the graph using BFS okie if a graph contain any loop or any cycle all element will nevere get traveresed because loop will go out of scope because topological sort means we have to print first those elements who come first**

```python
def solve(n,edges):
    adj_list=[[] for _ in range(n)]
    indegrees=[0]*n
    for a,b in edges:
        adj_list[a].append(b)
        indegrees[b]+=1
    
    from collections import deque
    queue=deque()
    result=[]
    for i in range(0,n):
        if indegrees[i]==0:
            queue.append(i)
    while len(queue)!=0:
        current_node=queue.popleft()
        result.append(current_node)
        for adjNode in adj_list[current_node]:
            indegrees[adjNode]-=1
            if indegrees[adjNode]==0:
                queue.append(adjNode)
    if len(result)!=n:
        return True
    else:
        return False
```
