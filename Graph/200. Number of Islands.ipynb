```python
def solve(board):
    """
    :type board: List[List[str]]
    :rtype: None Do not return anything, modify board in-place instead.
    """
    rows=len(board)
    cols=len(board[0])
    from collections import deque
    queue=deque()
    visited=[[0 for _ in range(cols)]for _ in range(rows)]

    for i in range(rows):
        for j in range(cols):
            if board[i][j]=="1":
                
                queue.append((i,j))
    def dfs(i,j,rows,cols,board,visited):
       
        if i<0 or i>=rows or j<0 or j>=cols:
            return 
        if board[i][j]=="1" and visited[i][j]==0:
            visited[i][j]=1
            dfs(i+1,j,rows,cols,board,visited)  #right movement
            dfs(i,j-1,rows,cols,board,visited)  #upward movement
            dfs(i-1,j,rows,cols,board,visited)  #left movement
            dfs(i,j+1,rows,cols,board,visited)  #downward movement
    count=0
    
    while len(queue)!=0:
        x,y=queue.popleft()
        import copy
        visited1=copy.deepcopy(visited)
        
        dfs(x,y,rows,cols,board,visited)
        
        flag=False
        if visited==board:
            return count
        
        for i in range(rows):
            for j in range(cols):
                
                if visited[i][j]!=visited1[i][j]:
                   
                   
                    flag=True
                    break
        if flag==True:
            count+=1



    # for i in range(0,rows):
    #     for j in range(0,cols):
    #         if visited[i][j]!=1:
    #             board[i][j]="X"
    
    return count
    
```
