```python
def solve(grid):
    import copy
    i,j=0,0
    fresh_orange=0
    row=len(grid)
    cols=len(grid[0])
    gridcopy=copy.deepcopy(grid)
    from collections import deque
    queue=deque()
    for i in range(row):
        for j in range(cols):
            if grid[i][j]==2:
                queue.append((i,j))
            elif grid[i][j]==1:
                fresh_orange+=1
            
    minutes=0
 
    while len(queue)!=0 and fresh_orange>0:
        
        minutes+=1
        print(gridcopy)
        total_rotten=len(queue)
        for _ in range(len(queue)):
            i,j=queue.popleft()
            for dx,dy in [(1,0),(-1,0),(0,1),(0,-1)]:
                new_i,new_j=i+dx,j+dy
                if new_i<0 or  new_i==row or new_j<0  or new_j==cols:
                    continue
                if gridcopy[new_i][new_j]==0 or gridcopy[new_i][new_j]==2:
                    continue
                fresh_orange-=1
                gridcopy[new_i][new_j]=2
                
                
                queue.append((new_i,new_j))
    if fresh_orange>0:
        return -1
   
    return minutes
                
                

            
```

```python
solve( [[2,1,1],[1,1,0],[0,1,1]])
```
