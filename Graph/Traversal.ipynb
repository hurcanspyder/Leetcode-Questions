```python
def solve(adjacencylist,n,startingnode):
    from collections import deque
    ans=[]
    queue=deque()
    visited=[0]*(n+1)
    queue.append(startingnode)
    visited[startingnode]=1
    while len(queue)!=0:
        e=queue.popleft()
        ans.append(e)
        for node in adjacencylist[e]:
            if visited[node]==0:
                queue.append(node)
                visited[node]=1
    return ans
```

```python
adjacencylist=[
    [],
    [2,8],
    [1,3,4],
    [2],
    [2,5],
    [4,6],
    [5,7],
    [6,8],
    [1,7,9],
    [8]
]
solve(adjacencylist,9,1)
```

**SC->O(3N)  TC->O(N)+O(2E)**
