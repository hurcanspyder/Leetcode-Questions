```python
class Solution(object):
    def isBipartite(self, graph):
        """
        :type graph: List[List[int]]
        :rtype: bool
        """
        def dfs(visited,graph,currentnode,color):
            visited[currentnode]=color
            for adjnode in graph[currentnode]:
                if visited[adjnode]!=-1:
                    if visited[adjnode]==color:
                        return False
                else:
                    if color==0:
                        ans=dfs(visited,graph,adjnode,1)
                    else:
                        ans=dfs(visited,graph,adjnode,0)
                    if ans==False:
                        return False
            return True


        n=len(graph)
        visited=[-1 for _ in range(n)]
        for index in range(0,n):
            if visited[index]==-1:
                ans=dfs(visited,graph,index,0)
                if ans==False:
                    return False
        return True
    
    

        
```
