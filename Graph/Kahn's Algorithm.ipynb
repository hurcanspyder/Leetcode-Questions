# Toposort

```python
def solve(n,edges):
    adj_list=[[] for _ in range(n)]
    indegrees=[0]*n
    for a,b in edges:
        adj_list[a].append(b)
        indegrees[b]+=1
    
    from collections import deque
    queue=deque()
    result=[]
    for i in range(0,n):
        if indegrees[i]==0:
            queue.append(i)
    while len(queue)!=0:
        current_node=queue.popleft()
        result.append(current_node)
        for adjNode in adj_list[current_node]:
            indegrees[adjNode]-=1
            if indegrees[adjNode]==0:
                queue.append(adjNode)
    return result
solve(8,[[1,2],[2,5],[2,3],[3,4],[5,6],[6,7]])
```

TC->O(n)+O(n)+O(V+2E)


SC->O(3n)
