```python
def solve(board):
    """
    :type board: List[List[str]]
    :rtype: None Do not return anything, modify board in-place instead.
    """
    rows=len(board)
    cols=len(board[0])
    from collections import deque
    queue=deque()
    visited=[[0 for _ in range(cols)]for _ in range(rows)]

    for i in range(rows):
        for j in range(cols):
            if board[i][j]=="O" and (i==0 or i==rows-1 or j==0 or j==cols-1):
                print(i,j)
                queue.append((i,j))
    def dfs(i,j,rows,cols,board,visited):
       
        if i<0 or i>=rows or j<0 or j>=cols:
            return 
        if board[i][j]=="O" and visited[i][j]==0:
            visited[i][j]=1


    while len(queue)!=0:
        x,y=queue.popleft()
        dfs(x,y,rows,cols,board,visited)
    for i in range(0,rows):
        for j in range(0,cols):
            if visited[i][j]!=1:
                board[i][j]="X"
   
    return board
```

```python
solve([["X","X","X","X"],["X","O","O","X"],["X","X","O","X"],["X","O","X","X"]])
```
