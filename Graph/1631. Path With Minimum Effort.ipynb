```python
class Solution(object):
    def minimumEffortPath(self, heights):
        """
        :type heights: List[List[int]]
        :rtype: int
        """
        rows=len(heights)
        cols=len(heights[0])
        effort=[[10**6 for _ in range(cols)]for _ in range(rows)]
        import heapq
        priority_queue=[]
        priority_queue.append([0,0,0])
        effort[0][0]=0
        while len(priority_queue)!=0:
            eff,i,j=heapq.heappop(priority_queue)
            if i==rows-1 and j==cols-1:
                return eff
            for x,y in [[1,0],[0,-1],[-1,0],[0,1]]:
                new_i,new_j=i+x,j+y
                if new_i<0 or new_i>=rows or new_j<0 or new_j>=cols:
                    continue
                next1=max(eff,abs(heights[new_i][new_j]-heights[i][j]))
                if next1<effort[new_i][new_j]:
                    effort[new_i][new_j]=next1
                    heapq.heappush(priority_queue,[next1,new_i,new_j])
```

TC->O(E LOGV)
