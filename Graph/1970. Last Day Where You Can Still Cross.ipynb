# Optimal Code

```python
class Solution(object):
    def latestDayToCross(self, row, col, cells):

        def dfs(r, c, grid, visited):
            if r < 0 or r >= row or c < 0 or c >= col:
                return False
            if grid[r][c] == 1 or visited[r][c] == 1:
                return False
            if r == row - 1:
                return True

            visited[r][c] = 1

            a = dfs(r + 1, c, grid, visited)
            if a: return True
            b = dfs(r, c - 1, grid, visited)
            if b: return True
            c2 = dfs(r, c + 1, grid, visited)
            if c2: return True
            d = dfs(r - 1, c, grid, visited)
            if d: return True

            return False

        # ----------------------------------------------------
        # binary search over days
        # ----------------------------------------------------

        left, right = 1, row * col
        ans = 1

        def can_cross(day):
            # build flooded grid for this day
            grid = [[0] * col for _ in range(row)]
            for i in range(day):
                r, c = cells[i]
                grid[r - 1][c - 1] = 1

            visited = [[0] * col for _ in range(row)]

            # try DFS from top row
            for j in range(col):
                if grid[0][j] == 0:
                    if dfs(0, j, grid, visited):
                        return True
            return False

        # binary search loop
        while left <= right:
            mid = (left + right) // 2
            if can_cross(mid):
                ans = mid
                left = mid + 1
            else:
                right = mid - 1

        return ans
```
