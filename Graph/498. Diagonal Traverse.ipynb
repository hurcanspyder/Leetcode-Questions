```python
class Solution(object):
    def findDiagonalOrder(self, mat):
        """
        :type mat: List[List[int]]
        :rtype: List[int]
        """
        from collections import deque

        queue=deque()
        for i in range(0,len(mat)):
            for j in range(0,len(mat[0])):
                if i==0 or j==len(mat[0])-1:
                    queue.append([i,j])
        def dfs(rows,cols,mat,list1):
            if rows<0 or rows>=len(mat) or cols<0 or cols>=len(mat[0]):
                return list1
            list1.append(mat[rows][cols])
            dfs(rows+1,cols-1,mat,list1)
            return list1
        list2=[]
        flag=True
        while len(queue)!=0:
            x,y=queue.popleft()
            a=dfs(x,y,mat,[])
            if flag:
                list2.append(a[::-1])
                flag=False
            else:
                list2.append(a)
                flag=True
        print(list2)
        list3=[]
        for i in range(len(list2)):
            for j in list2[i]:
                list3.append(j)
        return list3




        
```
