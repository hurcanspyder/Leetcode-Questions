```python
class Solution(object):
    def maxAreaOfIsland(self, grid):
        """
        :type grid: List[List[int]]
        :rtype: int
        """
        rows=len(grid)
        cols=len(grid[0])
        def func(i,j,grid,visited):
            if i<0 or i>=len(grid) or j<0 or j>=len(grid[0]):
                return 0
            if grid[i][j]==0 or visited[i][j]==1:
                return 0
            visited[i][j]=1
            left=func(i,j-1,grid,visited)
            right=func(i,j+1,grid,visited)
            up=func(i-1,j,grid,visited)
            down=func(i+1,j,grid,visited)
            return 1+left+right+up+down
        maxi=0
        visited=[[0 for _ in range(cols)]for _ in range(rows)]
        for i in range(rows):
            for j in range(cols):
                if grid[i][j]==1 and visited[i][j]==0:
                    maxi=max(maxi,func(i,j,grid,visited))
        return maxi
```
