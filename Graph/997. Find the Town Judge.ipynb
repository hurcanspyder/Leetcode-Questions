```python
class Solution(object):
    def findJudge(self, n, trust):
        adj_list = [[] for _ in range(n+1)]
        hash_table = {}

        for a, b in trust:
            adj_list[a].append(b)
            hash_table[b] = hash_table.get(b, 0) + 1

        for i in range(1, n+1):   # skip 0, since people are 1..n
            if adj_list[i] == [] and hash_table.get(i, 0) == n-1:
                return i

        return -1
```
