```python
class Solution(object):
    def openLock(self, deadends, target):
        """
        :type deadends: List[str]
        :type target: str
        :rtype: int
        """
        deadends=set(deadends)
        if "0000" in deadends:
            return -1
        from collections import deque
        queue=deque()
        queue.append("0000")
        count=0
        while len(queue)!=0:
            size=len(queue)
            for i in range(size):
                curr=queue.popleft()
                if curr==target:
                    return count
                for i in range(4):
                    digit=int(curr[i])

                    up=(digit+1)%10
                    next_combo=curr[:i]+str(up)+curr[i+1:]
                    if next_combo not in deadends:
                        queue.append(next_combo)
                        deadends.add(next_combo)
                    
                    down=(digit-1)%10
                    next_combo=curr[:i]+str(down)+curr[i+1:]
                    if next_combo not in deadends:
                        queue.append(next_combo)
                        deadends.add(next_combo)
            count+=1
        return -1
```
