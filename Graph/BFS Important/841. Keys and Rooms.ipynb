```python
class Solution(object):
    def canVisitAllRooms(self, rooms):
        """
        :type rooms: List[List[int]]
        :rtype: bool
        """
        my_set=set()
        my_set.add(0)
        visited=[0]*len(rooms)
        # for i in range(0,len(rooms)):
        #     if i in my_set:
        #         for j in rooms[i]:
        #             my_set.add(j)
        #     else:
        #         return False
        # return True
        from collections import deque
        queue=deque()
        queue.append(0)
        while len(queue)!=0:
            room=queue.popleft()
            visited[room]=1
            for diff_room in rooms[room]:
                if visited[diff_room]==0:
                    queue.append(diff_room)
        for i in visited:
            if i==0:
                return False
        return True


        
```
