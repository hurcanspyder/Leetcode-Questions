# Adjacency Matrix

```python
def adjacencymatrix(n,edges):
    m=n+1

    matrix=[[0 for _ in range(n+1)]for _ in range(n+1)]

    for u,v in edges:
        matrix[u][v]=1
        matrix[v][u]=1

    return matrix
    
```

```python
adjacencymatrix(5,[[1,2],[2,4],[3,4],[1,3],[3,5],[5,4]])
```

*Little bit costly take a lot of space*

# Adjacency List

*Less costly take space of O(2E) where E is no. of edges*

```python
def adjacencylist(n,edges):
    lst=[[] for _ in range(n+1)]

    for u,v in edges:
        lst[u].append(v)
        lst[v].append(u)
    return lst
```

```python
adjacencylist(5,[[1,2],[2,4],[3,4],[1,3],[3,5],[5,4]])
```

**adjacency list is Prefered**

# Dictonary 

*Not prefered because in general take less complexity but in Amortised case may be dictonary give TC of O(n)*

```python
def dictonary(n,edges):
    my_dict={}
    for i in range(1,n+1):
        my_dict[i]=[]
    for u,v in edges:
        my_dict[u].append(v)
        my_dict[v].append(u)
    return my_dict
```

```python
dictonary(5,[[1,2],[2,4],[3,4],[1,3],[3,5],[5,4]])
```
