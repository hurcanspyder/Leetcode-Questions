```python
class Solution(object):
    def uniquePathsIII(self, grid):
        
        rows = len(grid)
        cols = len(grid[0])
        
        empty = 0
        
        # Find start and count empty cells
        for i in range(rows):
            for j in range(cols):
                if grid[i][j] == 0:
                    empty += 1
                if grid[i][j] == 1:
                    start_i = i
                    start_j = j
        
        self.ans = 0
        
        def dfs(i, j, count):
            
           
            if i < 0 or j < 0 or i >= rows or j >= cols or grid[i][j] == -1:
                return
            
            
            if grid[i][j] == 2:
                if count == empty + 1:
                    self.ans += 1
                return
            
            temp = grid[i][j]
            grid[i][j] = -1  
            
            dfs(i+1, j, count+1)
            dfs(i-1, j, count+1)
            dfs(i, j+1, count+1)
            dfs(i, j-1, count+1)
            
            grid[i][j] = temp   
        
        dfs(start_i, start_j, 0)
        
        return self.ans


        
```
