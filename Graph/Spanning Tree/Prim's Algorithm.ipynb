*In queue we will save (weight,node,parent)*

*We are not gonna save 1 in place of visited here*

```python
def spanningtree(v,int,adjlist):
    import heapq
    mst=[]
    sum=0
    vis=[0 for _ in range(v)]
    priority_queue=[]
    priority_queue.append([0,0,-1]) #wt,node,parent
    while len(priority_queue)!=0:
        wt,node,parent=heapq.heappop(priority_queue)
        if vis[node]==0:
            vis[node]=1
            if parent!=-1:
                sum+=wt
                mst.append([parent,node])
            for adjNode,wt in adjlist[node]:
                if vis[adjNode]==0:
                    heapq.heappush(priority_queue,[wt,adjNode,node])
    return sum

```

TC->O(2E*logE)

**LogE is for heap finding minimum priority**
