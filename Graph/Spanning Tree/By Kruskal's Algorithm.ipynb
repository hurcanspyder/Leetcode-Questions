# BY Rank

```python
class DisjointSet:
    def __init__(self,n):
        self.parent=[i for i in range(0,n+1)]
        self.rank=[0]*(n+1)

    def find(self,x):
        if x==self.parent[x]:
            return x
        self.parent[x]=self.find(self.parent[x])
        return self.parent[x]

    def union(self,u,v):
        pu=self.find(u)
        pv=self.find(v)

        if pu==pv:
            return False

        if self.rank[pu]<self.rank[pv]:         #pu->pv
            self.parent[pu]=pv

        elif self.rank[pu]>self.rank[pv]:         #pv->pu
            self.parent[pv]=pu
        else:
            self.parent[pv]=pu
            self.rank[pu]+=1
        return True

```

```python
class Solution:
    def spanningtree(self,V,adj_list):
        edges=[]
        seen=set()
        # TC->O(N+E)

        for u in range(V):
            for neighbor in adj_list[u]:
                v,w=neighbor
                if (min(u,v) , max(u,v)) not in seen:
                    seen.add((min(u,v),max(u,v)))
                    edges.append((w,u,v))
        # TC->O(MlogM)
        edges.sort()
        dsu=DisjointSet(V)
        mst_weight=0
        # TC->O(M X 4 X alpha)
        for w,u,v in edges:
            if dsu.union(u,v):
                mst_weight+=w
            
        return mst_weight
```
