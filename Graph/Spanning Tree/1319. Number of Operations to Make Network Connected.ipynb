# Disjoitn Question

```python
class DisjointSet:
    def __init__(self,n):
        self.parent=[i for i in range(0,n+1)]
        self.rank=[0]*(n+1)

    def find(self,x):
        if x==self.parent[x]:
            return x
        self.parent[x]=self.find(self.parent[x])
        return self.parent[x]

    def union(self,u,v):
        pu=self.find(u)
        pv=self.find(v)

        if pu==pv:
            return True

        if self.rank[pu]<self.rank[pv]:         #pu->pv
            self.parent[pu]=pv

        elif self.rank[pu]>self.rank[pv]:         #pv->pu
            self.parent[pv]=pu
        else:
            self.parent[pv]=pu
            self.rank[pu]+=1
        return False
class Solution(object):
    def makeConnected(self, n, connections):
        """
        :type n: int
        :type connections: List[List[int]]
        :rtype: int
        """
        extra_edges=0
        ds=DisjointSet(n)
        for u,v in connections:
            if ds.union(u,v):
                extra_edges+=1
        
        components=0
        for i in range(n):
            if ds.find(i)==i:
                components+=1
        
        if extra_edges>=components-1:
            return components-1
        return -1


        
```
