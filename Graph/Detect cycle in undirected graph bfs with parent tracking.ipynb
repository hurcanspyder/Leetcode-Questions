```python
def isCycle(vertices,edges):
    adj_list=[[] for _ in range(vertices)]
    for u,v in edges:
        adj_list[u].append(v)
        adj_list[v].append(u)
    vis=[0]*vertices

    from collections import deque
    queue=deque()
    queue.append((0,-1))
    vis[0]=1
    for i in range(0,vertices):
        if vis[i]==1:
            continue
        while len(queue)!=0:
            node,parent=queue.popleft()
            for adjnode in adj_list[node]:
                if vis[adjnode]==0:
                    vis[adjnode]=1
                    queue.append((adjnode,node))
                else:
                    if adjnode!=parent:
                        return True
    return False
```
