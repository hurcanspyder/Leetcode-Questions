Question	BFS	DFS

Level-wise	✅	❌

4️⃣ Graph Weights Decide Everything 🔥
❌ Weighted Graph?

DFS → ❌

BFS → ❌ (if weights differ)

✅ What to use?

Dijkstra → positive weights

Bellman-Ford → negative weights
