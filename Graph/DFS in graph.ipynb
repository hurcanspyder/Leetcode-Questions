```python
def dfs(node,result,visited,adjacencylist):
    visited[node]=1
    result.append(node)
    for n in adjacencylist[node]:
        if visited[n]==0:
            dfs(n,result,visited,adjacencylist)
    return result
```

```python
adjacencylist=[
    [],
    [2,8],
    [1,3,4],
    [2],
    [2,5],
    [4,6],
    [5,7],
    [6,8],
    [1,7,9],
    [8]

]
adj=[[0], [1], [2]]
visited=[0]*(2+1)
dfs(0,[],visited,adj)
```

**SC->O(2N)+O(Depth) and TC->O(N)+O(2E)**
