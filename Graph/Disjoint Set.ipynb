# Union by Rank

```python
class DisjointSet:
    def __init__(self,n):
        self.parent=[i for i in range(0,n+1)]
        self.rank=[0]*(n+1)

    def find(self,x):
        if x==self.parent[x]:
            return x
        self.parent[x]=self.find(self.parent[x])
        return self.parent[x]

    def union(self,u,v):
        pu=self.find(u)
        pv=self.find(v)

        if pu==pv:
            return 

        if self.rank[pu]<self.rank[pv]:         #pu->pv
            self.parent[pu]=pv

        elif self.rank[pu]>self.rank[pv]:         #pv->pu
            self.parent[pv]=pu
        else:
            self.parent[pv]=pu
            self.rank[pu]+=1
```

# Union BY Size

```python
class DisjointSet:
    def __init__(self,n):
        self.parent=[i for i in range(0,n+1)]
        self.size=[1]*(n+1)

    def find(self,x):
        if x==self.parent[x]:
            return x
        self.parent[x]=self.find(self.parent[x])
        return self.parent[x]

    def union(self,u,v):
        pu=self.find(u)
        pv=self.find(v)

        if pu==pv:
            return 

        if self.size[pu]<self.size[pv]:         #pu->pv
            self.parent[pu]=pv
            self.size[pv]+=self.size[pu]

        elif self.size[pu]>self.size[pv]:         #pv->pu
            self.parent[pv]=pu
            self.size[pu]+=self.size[pv]
        else:
            self.parent[pv]=pu
            self.size[pu]+=self.size[pv]
```
