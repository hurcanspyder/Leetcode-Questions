# Dijkstra's ALGORITHM using Priority Queue(heap)

```python
import heapq
def dijkstras(v,edges,src):
    adj_list=[[] for _ in range(v)]
    for u,v,d in edges:
        adj_list[u].append([v,d])
    distance=[float("inf") for _ in range(v)]
    distance[src]=0

    priority_queue=[[0,src]]

    while len(priority_queue)!=0:
        curr_dist,node=heapq.heappop(priority_queue)
        if curr_dist>distance[node]:
            continue
        for adjNode,weight in adj_list[node]:
            dist_trav=curr_dist+weight
            if dist_trav<distance[node]:
                distance[node]=dist_trav
                priority_queue.append([dist_trav,adjNode])
    return distance
```

**This will only work if**

There is no negative weight

There is no negative cycle 

TC->E log(V)



*We can do this same code using set but it will get more complex where we are using heap we will use set instead by this we will get more complexity beacuse to find the minimum weight we have to iterate it all which gives us more complex TC which is O(V)*

# why we use priority queue over simple queue

*In queue we have to iterate for all present weights and adjacent node means we have to look for all the combinatiosn present in our graph which automatically increase the complexity of the code but where as in priority queue we have to look for the smallest weight which give us good complexity*
