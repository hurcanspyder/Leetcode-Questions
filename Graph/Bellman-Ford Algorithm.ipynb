**Total iterations will be n-1 cause**

PROOF:-

check for 0->1->2->3->4->5

*here we have to relax all the elements present in the grph*

**Like This if [u,v,distance] then if (d[u]+distance<d[v] then d[v]=d[u]+distance)**

```python
class Solution:
    def bellmanFord(self, V, edges, src):
        INF = 10**8                       # a very large number
        dist = [INF for _ in range(V)]
        dist[src] = 0                     # distance to source is zero

        # repeat edge relaxation V-1 times
        for _ in range(V - 1):
            for u, v, w in edges:
                if dist[u] != INF and dist[u] + w < dist[v]:
                    dist[v] = dist[u] + w

        # check for negative-weight cycles
        for u, v, w in edges:
            if dist[u] != INF and dist[u] + w < dist[v]:
                return [-1]               # cycle found

        return dist                       # shortest distances
```
