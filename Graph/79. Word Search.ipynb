```python
class Solution(object):
    def exist(self, board, word):
        """
        :type board: List[List[str]]
        :type word: str
        :rtype: bool
        """
        rows=len(board)
        cols=len(board[0])
        visited=[[0 for _ in range(cols)]for _ in range(rows)]
        from collections import deque
        queue=deque()
        for i in range(rows):
            for j in range(cols):
                if board[i][j]==word[0]:
                    queue.append((i,j))
                    
        flag=0
        def dfs(i,j,rows,cols,word,index,visited,board,flag):
            if index==len(word)-1:
                flag=1

            if i<0 or i>=rows or j<0 or j>=cols or index>=len(word):
                return 

            
        
            if board[i][j]==word[index] and visited[i][j]==0:
                visited[i][j]=1
                dfs(i+1,j,rows,cols,word,index+1,visited,board,flag)  #right movement
                dfs(i,j-1,rows,cols,word,index+1,visited,board,flag)  #upward movement
                dfs(i-1,j,rows,cols,word,index+1,visited,board,flag)  #left movement
                dfs(i,j+1,rows,cols,word,index+1,visited,board,flag)  #downward movement
            
        while len(queue)!=0:
            visited=[[0 for _ in range(cols)]for _ in range(rows)]
            x,y=queue.popleft()
            dfs(x,y,rows,cols,word,0,visited,board,flag)
            print(visited)
            return flag
            
          
        




        

        
```
