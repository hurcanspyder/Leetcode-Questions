```python
class Solution(object):
    def canFinish(self, numCourses, prerequisites):
        vertices = numCourses
        edges = prerequisites

        def dfs(currentnode, visited, path_visited, adj_list):
            visited[currentnode] = 1
            path_visited[currentnode] = 1

            for adjNode in adj_list[currentnode]:
                if visited[adjNode] == 0:
                    if dfs(adjNode, visited, path_visited, adj_list):
                        return True
                elif path_visited[adjNode] == 1:
                    return True

            path_visited[currentnode] = 0
            return False

        adj_list = [[] for _ in range(vertices)]
        visited = [0] * vertices
        path_visited = [0] * vertices

        for a, b in edges:   # b → a (to take course a, must finish b)
            adj_list[b].append(a)

        for i in range(vertices):
            if visited[i] == 0:
                if dfs(i, visited, path_visited, adj_list):
                    return False   # cycle detected → cannot finish

        return True   # no cycles → can finish
```
