```python
class Solution(object):
    def findWords(self, board, words):
        rows = len(board)
        cols = len(board[0])

        def dfs(r, c, index, visited, word, board, result):
            # If full word found
            if index == len(word):
                if word not in result:
                    result.append(word)
                return

            # Out of bounds
            if r < 0 or r >= rows or c < 0 or c >= cols:
                return

            # Already used
            if visited[r][c] == 1:
                return

            # Character mismatch
            if board[r][c] != word[index]:
                return

            # Mark visited
            visited[r][c] = 1

            # Explore 4 directions
            dfs(r+1, c, index+1, visited, word, board, result)
            dfs(r-1, c, index+1, visited, word, board, result)
            dfs(r, c+1, index+1, visited, word, board, result)
            dfs(r, c-1, index+1, visited, word, board, result)

            # Backtrack
            visited[r][c] = 0

        final = []
        from collections import deque
        
        for word in words:
            queue = deque()
            
            # Find all starting positions
            for i in range(rows):
                for j in range(cols):
                    if board[i][j] == word[0]:
                        queue.append((i, j))

            # Try DFS from every starting point
            while queue:
                visited = [[0]*cols for _ in range(rows)]
                x, y = queue.popleft()
                dfs(x, y, 0, visited, word, board, final)

        return final
```
