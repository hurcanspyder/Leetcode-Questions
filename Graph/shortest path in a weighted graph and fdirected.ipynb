```python
adj_list=[[[3,3]],[[3,1]],[[1,2]],[[[]]],[[0,1],[2,3]],[[4,3],[6,2]],[[4,1]]]
```

```python
def toposort(edges,n,src):
    def dfs(node,stack,visited,adj_list):
        visited[node]=1
        for adjNode,distance in adj_list[node]:
            if visited[adjNode]==0:
                dfs(adjNode,stack,visited,adj_list)
        stack.append(node)
    # from collections import deque
    # queue=deque()
    # indegrees=[0]*n
    # result=[]
    adj_list=[[]for _ in range(n)]
    for u,v,d in edges:
        adj_list[u].append([v,d])
    stack=[]
    visited=[0]*n
    for i in range(0,n):
        if visited[i]==0:
            dfs(i,stack,visited,adj_list)
    distance=[10*20]*n
    distance[src]=0
    print(stack)
    while len(stack)!=0:
        node=stack.pop()
        dist=distance[node]
        for adjNode,d in adj_list[node]:
            new_d=dist+d
            if new_d<distance[adjNode]:
                distance[adjNode]=new_d
    for i in range(0,n):
        if distance[i]==10*20:
            distance[i]=-1
    return distance


```
