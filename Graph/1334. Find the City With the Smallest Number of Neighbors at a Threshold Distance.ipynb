```python
class Solution(object):
    def findTheCity(self, n, edges, distanceThreshold):
        """
        :type n: int
        :type edges: List[List[int]]
        :type distanceThreshold: int
        :rtype: int
        """
        adj_matrix=[[10**8 for _ in range(n)]for _ in range(n)]
        for i,j,dist in edges:
            adj_matrix[i][j]=dist
            adj_matrix[j][i]=dist
            adj_matrix[i][i]=0
            adj_matrix[j][j]=0
        
        
        for via in range(n):
            for i in range(n):
                for j in range(n):
                    if adj_matrix[i][via] !=10**8 and adj_matrix[via][j] != 10**8:
                        adj_matrix[i][j] = min(adj_matrix[i][j],
                                         adj_matrix[i][via] + adj_matrix[via][j])
        list1=[0 for _ in range(n)]
        min1=10**8
        min_index=0
        for i in range(n):
            for j in range(n):
                if adj_matrix[i][j]<=distanceThreshold and i!=j:
                    list1[i]+=1
            if list1[i]<=min1:
                min1=list1[i] 
                min_index=i
        return min_index 
                
        


        
```
