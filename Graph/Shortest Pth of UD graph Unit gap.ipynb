```python
def shortestpath(n,adj_list):
    from collections import deque
    queue=deque()
    visited=[-1]*n
    queue.append([0,0])
    count=0
    visited[0]=0
    while len(queue)!=0:
        a,b=queue.popleft()
        

        for node in adj_list[a]:
            if visited[node]==-1:
                visited[node]=b+1
                queue.append([node,b+1])
            # else:
            #     visited[node]=min(visited[node],count)
    return(visited)
        

```
