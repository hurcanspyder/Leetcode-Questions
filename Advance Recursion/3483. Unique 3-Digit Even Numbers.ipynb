```python
def solve(nums,subset,index,result):
    if len(subset)==3:
        return result.append(subset.copy())
        

    if index>=len(nums):
        return 
    subset.append(nums[index])
    solve(nums,subset,index+1,result)
    subset.pop()
    solve(nums,subset,index+1,result)
    return result
   
```

```python
solve([1,2,3],[],0,[])
```
