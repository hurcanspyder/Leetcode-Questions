If we don't sort, recursion explores values in different orders:

We sort nums so that:

duplicates stand next to each other, making it possible to avoid generating duplicate subsets before they even appear.

In Python, an object is hashable if:

✔ it has a hash value that doesn't change during its lifetime

✔ it can be used as a key in a dictionary

✔ it can be stored in a set

| Object  | Hashable? | Why       |
| ------- | --------- | --------- |
| list    | ❌         | mutable   |
| dict    | ❌         | mutable   |
| set     | ❌         | mutable   |
| tuple   | ✔️        | immutable |
| int/str | ✔️        | immutable |


```python
class Solution(object):
    def subsetsWithDup(self, nums):
        """
        :type nums: List[int]
        :rtype: List[List[int]]
        """
        nums.sort()
        def solve(nums,index,subset,result):
            
            if index>len(nums):
                return
            if index==len(nums):
                result.add(tuple(subset[:]))
                return result
            subset.append(nums[index])
            solve(nums,index+1,subset,result)
            subset.pop()
            solve(nums,index+1,subset,result)
            return result
        self.result=set()
        solve(nums,0,[],self.result)
        return [list(t) for t in self.result]

            
            
        
```
