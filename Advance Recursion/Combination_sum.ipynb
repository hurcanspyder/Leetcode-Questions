```python
def combinationsum(subset,total,index,nums,target,result):
    # if sum(nums)<target:
    #     return -1
    if target==0:
        return 0
    if total==target:
        result.append(subset.copy())
        return
    if index>=len(nums):
        return
    if total>target:
        return 
    subset.append(nums[index])
    sum1=total+nums[index]
    combinationsum(subset,sum1,index,nums,target,result)
    subset.pop()
    sum1=sum1-nums[index]
    combinationsum(subset,sum1,index+1,nums,target,result)
    return result



```

```python
combinationsum([],0,0,[1,4,9],12,[])
```

```python
a=combinationsum([],0,0,[2],3,[])
if a==[]:
    return -1

b=[]
for i in a:
    b.append(len(i))
min_index=b.index(min(b))
print(a[min_index])
```
