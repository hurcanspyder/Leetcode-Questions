```python
def solve(ind,subset):
    nums=[8,5,7]
    
    if ind>=len(nums):
    
        return [subset.copy()]
    subset.append(nums[ind])
    left=solve(ind+1,subset)
    
    subset.pop()
    right=solve(ind+1,subset)
   
    
    return left+right


    

```
