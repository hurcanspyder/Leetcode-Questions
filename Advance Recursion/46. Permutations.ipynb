```python
def solve(subset,result,index,nums):
    
    if len(subset)==3:
        return result.append(subset[:])
    if index==len(nums):
        return 
    subset.append(nums[index])
    solve(subset,result,index+1,nums)
    subset.pop()
    solve(subset,result,index+1,nums)
    return result
```

```python
solve([],[],0,[1,2,3])
```
