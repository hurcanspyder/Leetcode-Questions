```python
class Solution(object):
    def solveNQueens(self, n):
        """
        :type n: int
        :rtype: List[List[str]]
        """
        grid=[["." for _ in range(n)]for _ in range(n)]
        def left(i,j,grid):
            
            if j<0:
                return True
            if grid[i][j]=="Q":
                
                return False
            return left(i,j-1,grid)
        
        def up(i,j,grid):
            if i<0 or j<0:
                return True
            if grid[i][j]=="Q":
                return False
            return up(i-1,j-1,grid)
        
        def bottom(i,j,grid):
            if i>=len(grid) or j<0:
                return True
            if grid[i][j]=="Q":
                return False
            return bottom(i+1,j-1,grid)
        
        self.result=[]
        
        def dfs(i,j,grid):
            if j == n:
                self.result.append(["".join(r) for r in grid])  # FIX 2: deep copy
                return
            if i>=n:
                
                return 

            if left(i,j,grid) and up(i,j,grid) and bottom(i,j,grid):
                grid[i][j]="Q"
                dfs(0,j+1,grid)
                grid[i][j]="."
            dfs(i+1,j,grid)
        dfs(0,0,grid)
        return self.result
```
