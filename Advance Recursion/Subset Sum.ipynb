```python
def subsetsum(index,subset,result,total,nums):
    if index>=len(nums):
        result.append(total)
        return 
    subset.append(nums[index])
    sum=total+nums[index]
    subsetsum(index+1,subset,result,sum,nums)
    subset.pop()
    sum=sum-nums[index]
    subsetsum(index+1,subset,result,total,nums)
    return result

```
