```python
def combination(digits):
    dict1={'2':'abc','3':'def','4':'ghi','5':'jkl','6':'mno','7':'pqrs','8':'tuv','9':'wxyz'}
    string=''
    
    for i in digits:
        string=string+dict1[i]
    def lettercombination(string,subset,n,result,index):
        
           
        if len(subset)==n:
            
            result.append(''.join(subset[:]))
            return
        if index>=len(string):
            return 
        subset.append(string[index])    
        lettercombination(string,subset,n,result,index+1)
        subset.pop()
        lettercombination(string,subset,n,result,index+1)
        return result
    a=lettercombination(string,[],len(digits),[],0)
    return a
    


```

```python
combination("23")
```
