```python
def generate(subset,index,total,result):
    
    
    if index>=len(subset):
        if total==0:
            result.append("".join(subset))
        return
    if total>len(subset)//2: return 
    elif total<0: return
        


    
    subset[index]="("
    sum=total+1
    generate(subset,index+1,sum,result)
    
    subset[index]=")"
    sum=total-1
    generate(subset,index+1,sum,result)
    return result    
    
    
    

    
```

```python
generate(["","","",""],0,0,[])
```

**Time complexity for this is O(2^n) and stack space is O(2n)**
