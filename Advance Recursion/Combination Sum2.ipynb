# Final valid code

**Pruning:** Skiping all elements which are greater then target

```python
class Solution(object):
    def combinationSum2(self, candidates, target):
        candidates.sort()
        result = set()

        def func(index, subset, target):
            if target == 0:
                result.add(tuple(subset))
                return
            
            if index == len(candidates):
                return
            
            for i in range(index, len(candidates)):
                if i > index and candidates[i] == candidates[i - 1]:
                    continue
                if candidates[i] > target:          #Pruning
                    break
                
                subset.append(candidates[i])
                func(i + 1, subset, target - candidates[i])
                subset.pop()
        
        func(0, [], target)
        return [list(r) for r in result]
```
