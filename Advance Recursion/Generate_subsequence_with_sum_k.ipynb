```python
def backtrack(index,total,subset,target,nums,result):
    if total==target:
        return result.append(subset.copy())
    elif total>target:
        return
    if index>=len(nums):
        return
    subset.append(nums[index])
    sum=total+nums[index]
    backtrack(index+1,sum,subset,target,nums,result)
    e=subset.pop()
    sum=sum-e
    backtrack(index+1,sum,subset,target,nums,result)
    return len(result)
    
```
