```python
def backtrack(index,total,subset,target,nums,result):
    if total==target:
        return result.append(subset.copy())
    elif total>target:
        return
    if index>=len(nums):
        return
    subset.append(nums[index])
    sum=total+nums[index]
    backtrack(index+1,sum,subset,target,nums,result)
    e=subset.pop()
    sum=sum-e
    backtrack(index+1,sum,subset,target,nums,result)
    return len(result)
    
```

```python
backtrack(0,0,[],3,[1,3,2,1],[])
```

```python
def rotate(nums, k):
    """
    :type nums: List[int]
    :type k: int
    :rtype: None Do not return anything, modify nums in-place instead.
    """
    if k%len(nums)==0:
        return nums
    if k<len(nums):
        return nums[len(nums)-k:]+nums[:len(nums)-k]
```

```python
rotate([1,2,3,4,5,6,7],3)
```
