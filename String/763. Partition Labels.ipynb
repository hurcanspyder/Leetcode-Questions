```python
class Solution(object):
    def partitionLabels(self, s):
        """
        :type s: str
        :rtype: List[int]
        """
        left=0
        my_set=set()
        my_set.add(s[0])
        list1=[]
        for i in range(1,len(s)):
            flag=True
            for j in my_set:
                if j in s[i:]:
                    flag=False
            if flag:
                list1.append(i-left)
                left=i
            if s[i] not in my_set:
                my_set.add(s[i])
        list1.append(len(s)-left)
        return list1


            
        
```
