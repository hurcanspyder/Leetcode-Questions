```python
class Solution(object):
    def removeOuterParentheses(self, s):
        """
        :type s: str
        :rtype: str
        """
        count=0
        p=""
        for i in s:
            if i=="(":
                
                if count!=0:
                    p=p+i
                count+=1
                
            if i==")":
                count-=1
                if count!=0:
                    p=p+i
                
            
        return p
            
```
