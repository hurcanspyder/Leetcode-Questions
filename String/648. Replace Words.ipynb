```python
class Solution(object):
    def replaceWords(self, dictionary, sentence):
        """
        :type dictionary: List[str]
        :type sentence: str
        :rtype: str
        """
        dict1=set(dictionary)
        final_str=""
        list1=[]
        str1=""
        for i in range(0,len(sentence)):
            
            if sentence[i]!=" ":
                str1+=sentence[i]
            else:
                if len(str1)!=0:
                    list1.append(str1)
                    str1=""
        list1.append(str1)
        for i in range(0,len(list1)):
            str2=""
            for j in range(0,len(list1[i])):
                str2+=list1[i][j]
                if str2 in dict1:
                    list1[i]=str2
                    break
        return " ".join(list1)
```
