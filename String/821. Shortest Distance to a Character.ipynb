```python
class Solution(object):
    def shortestToChar(self, s, c):
        """
        :type s: str
        :type c: str
        :rtype: List[int]
        """
        list1=[]
        for i in range(0,len(s)):
            if s[i]==c:
                list1.append(i)
        list2=[]
        for i in range(0,len(s)):
            mini=10**8
            for j in list1:
                mini=min(mini,abs(j-i))
            list2.append(mini)
        return list2
        
```
