```python
class Solution(object):
    def customSortString(self, order, s):
        """
        :type order: str
        :type s: str
        :rtype: str
        """
        hash_map={}
        for i in s:
            if i not in hash_map:
                hash_map[i]=1
            else:
                hash_map[i]+=1
        list1=[]
        for i in order:
            if i in hash_map:
                a=[i]*hash_map[i]
                list1+=a
                hash_map.pop(i)
        for key,value in hash_map.items():
            a=[key]*value
            list1+=a
        return "".join(list1)
        


        
```
