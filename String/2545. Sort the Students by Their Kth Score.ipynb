```python
class Solution(object):
    def sortTheStudents(self, score, k):
        """
        :type score: List[List[int]]
        :type k: int
        :rtype: List[List[int]]
        """
        hash_map={}
        for i in range(0,len(score)):
            hash_map[i]=score[i][k]
        sorted1=sorted(hash_map.items(),key=lambda x:x[1],reverse=True)
        list2=[]
        for key,value in sorted1:
            list2.append(score[key])
        return list2
        
```
