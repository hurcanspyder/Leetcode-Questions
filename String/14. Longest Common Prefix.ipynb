```python
class Solution(object):
    def longestCommonPrefix(self, strs):
        """
        :type strs: List[str]
        :rtype: str
        """
        s=""
        if len(strs)==0:
            return ""
        for i in range(0,len(strs[0])):
            for j in range(1,len(strs)):
                if i==len(strs[j]) or strs[j][i]!=strs[0][i]:
                    return s
            s+=strs[0][i]
        return s   
        
```
