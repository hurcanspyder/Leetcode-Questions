```python
class Solution(object):
    def twoEggDrop(self, n):
        """
        :type n: int
        :rtype: int
        """
        if n<=2:
            return n
        list1=[]
        count=1
        while n>0:
            n-=count
            list1.append(n)
            count+=1
        list1.pop()
        return abs(list1[-1]-list1[-2])+1

        
```
