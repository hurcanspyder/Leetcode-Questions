# This is mmore optimal rather then DFS or BFS

```python
class Solution(object):
    def escapeGhosts(self, ghosts, target):
        tx, ty = target

        # Your distance from (0,0)
        my_dist = abs(tx) + abs(ty)

        # Check each ghost
        for gx, gy in ghosts:
            ghost_dist = abs(gx - tx) + abs(gy - ty)
            if ghost_dist <= my_dist:
                return False

        return True
```
