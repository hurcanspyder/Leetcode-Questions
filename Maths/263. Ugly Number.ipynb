*Here the core concept is that each and every positive number is a multiple of prime number its an theory in a mathematics that each positive number can be written in multiple of prime numbers*

```python
class Solution(object):
    def isUgly(self, n):
        if n <= 0:
            return False

        for d in [2, 3, 5]:
            while n % d == 0:
                n //= d
        
        return n == 1
```
