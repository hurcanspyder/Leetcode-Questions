```python
class Solution(object):
    def maxChunksToSorted(self, arr):
        """
        :type arr: List[int]
        :rtype: int
        """
        expected_sum,chunk,actual_sum=0,0,0
        for i in range(0,len(arr)):
            actual_sum+=arr[i]
            expected_sum+=i
            if actual_sum==expected_sum:
                chunk+=1
        return chunk

    
        
```
