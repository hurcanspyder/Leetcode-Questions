```python
class Solution(object):
    def constructArray(self, n, k):
        """
        :type n: int
        :type k: int
        :rtype: List[int]
        """
        if k==1:
            return [i for i in range(1,n+1)]
        list1=[]
        count=1
        count1=0
        flag=True
        for i in range(1,k+1):
            if flag:
                list1.append(count)
                count+=1
                flag=False
            else:
                list1.append(k+1-count1)
                count1+=1
                flag=True
        my_set=set(list1)
        for i in range(1,n+1):
            if i not in my_set:
                list1.append(i)
        return list1
```
