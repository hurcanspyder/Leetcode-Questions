```python
class Solution(object):
    def complexNumberMultiply(self, num1, num2):
        """
        :type num1: str
        :type num2: str
        :rtype: str
        """
        index1=0
        index2=0
        for i in range(len(num1)):
            if num1[i]=="+":
                index1=i
        for j in range(len(num2)):
            if num2[j]=="+":
                index2=j
        


        
        real1=num1[:index1]
        real2=num2[:index2]
        imaginary1=num1[index1+1:-1]
        imaginary2=num2[index2+1:-1]
        product=str(int(real1)*int(real2)-int(imaginary1)*int(imaginary2)) + "+" + \
str(int(real1)*int(imaginary2)+int(real2)*int(imaginary1)) + "i"

        return product
```
