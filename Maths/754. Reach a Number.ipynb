# Recursive approach very inefficient

```python
class Solution(object):
    def reachNumber(self, target):
        """
        :type target: int
        :rtype: int
        """
        def func(dig,target,sum1):
            if sum1==target:
                return 0
            if sum1>target:
                return 10**8
            if sum1<-target:
                return 10**8
            new_dig=dig+1
            right=1+func(new_dig,target,sum1+new_dig)
            left=1+func(new_dig,target,sum1-new_dig)
            return min(right,left)
        return func(0,target,0)
        


            

```

# optimal way

```python
class Solution(object):
    def reachNumber(self, target):
        target = abs(target)
        sum_steps, num_moves = 0, 0

        while sum_steps < target or (sum_steps - target) % 2 != 0:
            num_moves += 1
            sum_steps += num_moves

        return num_moves
```
