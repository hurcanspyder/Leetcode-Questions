```python
class Solution(object):
    def findGCD(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        max_ele=max(nums)
        min_ele=min(nums)
        if max_ele%min_ele==0:
            return min_ele
        else:
            def func(maxi,mini):
                rem=maxi%mini
                if rem==0:
                    return mini
                return func(mini,rem)
                
            return func(max_ele,min_ele)
        

        
```
