```python
class Solution(object):
    def maxChunksToSorted(self, arr):
        """
        :type arr: List[int]
        :rtype: int
        """
        sorted1=sorted(arr)
        chunk=0
        for i in range(0,len(arr)):
            new_arr=arr[:i+1]
            new_arr.sort()
            if new_arr==sorted1[:i+1]:
                chunk+=1
        return chunk
        
```
