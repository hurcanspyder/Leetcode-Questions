```python
class Solution(object):
    def productExceptSelf(self, nums):
        """
        :type nums: List[int]
        :rtype: List[int]
        """
        product=1
        my_dict={}
        for i in range(1,len(nums)):
            product*=nums[i]
        for i in range(0,len(nums)):
            if nums[i] not in my_dict:
                my_dict[nums[i]]=1
            else:
                my_dict[nums[i]]+=1
        if 0 in my_dict:
            if my_dict[0]>1:
                return [0]*len(nums)
        
        if product==0:
            product_zero=1
            for i in range(0,len(nums)):
                if nums[i]!=0:
                    product_zero=product_zero*nums[i]
                
                
                    
        list1=[]
        list1.append(product)
        
        for i in range(1,len(nums)):
            if nums[i]!=0:
                product=(list1[-1]/nums[i])*nums[i-1]
            else:
                product=product_zero
            list1.append(product)
        return list1
        
```
