```python
def canPlaceFlowers(flowerbed, n):
    """
    :type flowerbed: List[int]
    :type n: int
    :rtype: bool
    """
    if flowerbed==[0] and n==1:
        return True

    count=0
    for i in range(0,len(flowerbed)):
        if flowerbed[i]==0:
            for j in range(i+1,len(flowerbed)):
                print(i)
                if i==0 and j==1 and flowerbed[j]==0:
                    print("a")
                    count+=1
                    i=j
                if j==len(flowerbed)-1 and flowerbed[j]==0:
                    print("b",i,j)
                    count+=1
                
                if j==i+2 and flowerbed[j]==0:
                    print("c",i,j)
                    
                    count+=1
                    i=i+2
                    
                if flowerbed[j]!=0:
                    
                    break
    
    return count>=n
```

```python
canPlaceFlowers([1,0,0,0,0,1],2)
```
