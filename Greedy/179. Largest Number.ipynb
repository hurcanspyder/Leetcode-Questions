```python
class Solution(object):
    def largestNumber(self, nums):
        """
        :type nums: List[int]
        :rtype: str
        """
        my_set=set()
        list1=[]
        list2=[]
        if len(nums)==1:
            return str(nums[0])
        for i in range(len(nums)):
            nums[i]=str(nums[i])
        while len(list1)!=len(nums):
            max_element="0"
            for index,value in enumerate(nums):
                if index in list2:
                    print("Hello")
                    continue
                else:
                    
                    if int(value[0])>int(max_element[0]):
                        max_element=value
                        # print(my_set)
                        # if my_set:
                        #     my_set.remove(last_index)
                        # my_set.add(index)

                        last_index=index
                    if int(value[0])==int(max_element[0]):
                        if int(max_element+value)<=int(value+max_element):
                            max_element=value
                            # if my_set:
                            #     my_set.remove(last_index)
                            # my_set.add(index)
                            last_index=index
            
        
                        
                        

            list2.append(last_index)    
            list1.append(max_element)
        if list1[0]=="0":
            return "0"
        return "".join(list1)
                    
        
```
