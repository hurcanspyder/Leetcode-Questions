```python
def solve(nums,index):
    target=len(nums)-1
    list1=[]
    list1.append(nums[index])
    if sum(list1)==target:
        return len(list1)
    
```

```python
def solve1(nums):
    jump=0
    left=0
    right=0
    n=len(nums)
    while right<n-1:
        farthest=0
        for i in range(left,right+1):
            farthest=max(farthest,i+nums[i])
        left=right+1
        right=farthest
        jump+=1
    return jump

```
