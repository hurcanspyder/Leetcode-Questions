```python
def solve(matrix,target):
    total=0
    value=0
    for i in range(0,len(matrix)):
        if total==target:
            return value
        if total+matrix[i][1]>target:
            value+=(matrix[i][0]/matrix[i][1])*(target-total)
            total=target
            
            
        else:
            total+=matrix[i][1]
            value+=matrix[i][0]
            
            print(value)
    return value


```

```python
solve([[100,10],[200,50],[300,100],[50,10]],90)
```
