```python
def solve(arr,dep):
    left=1
    platform=1
    count=0
    max_platform=0
    while left<len(arr):
        # if left==0:
        #     platform+=1
        if arr[left]>dep[left-1]:
            count=0
        if arr[left]<dep[left-1]:
            count+=1
        
            if count==0:
                platform=0
                platform+=1
            else:
                platform+=1
        print(platform,left)
        max_platform=max(max_platform,platform)
        left+=1
    return max_platform
```
