```python
def solve(g,s):
    n=len(g)
    m=len(s)
    g.sort()
    s.sort()

    i=0
    j=0
    count=0
    while i<n and j<m:
        if g[i]<=s[j]:
            count+=1
            j+=1
            i+=1
        else:
            j+=1
    return count
            
            
    
        
```

```python
solve([1,2],[1,2,3])
```
